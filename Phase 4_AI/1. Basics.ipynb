# Artificial Intelligence (AI) — BASICS

## 🔹 Basic Theory (Beginner-friendly)

Artificial Intelligence (AI) is the field of computer science that focuses on building machines that can think, reason, learn, and act like humans.

### In simple words:

AI = Making computers intelligent enough to solve problems by themselves

### Examples:

- Google Maps finding the shortest path  
- ChatGPT answering questions  
- A robot avoiding obstacles  
- Chess-playing programs  

### AI systems can:

- Perceive (see, hear, sense)  
- Reason (think logically)  
- Learn (improve from experience)  
- Act (take decisions)  

## 🔹 What AI is NOT

- AI is not only Machine Learning  
- AI existed before ML and DL  
- ML/DL are tools inside AI

# Artificial Intelligence (AI) — ADVANCED THEORY

## 🔹 Formal Definition (Exam / Interview)

Artificial Intelligence is the study and design of intelligent agents that perceive their environment and take actions that maximize their chances of achieving goals.

(Russell & Norvig)

## 🔹 4 Classical Approaches to AI

### 1️⃣ Acting Humanly (Turing Test)

If a machine behaves like a human → it is intelligent

Example:
- Chatbots

### 2️⃣ Thinking Humanly (Cognitive Modeling)

- Mimic human brain processes  
- Used in psychology & neuroscience  

### 3️⃣ Thinking Rationally (Laws of Thought)

- Logical reasoning  
- Mathematical proofs, logic systems  

### 4️⃣ Acting Rationally ✅ (MOST IMPORTANT)

- Agent takes the best possible action  
- Modern AI focuses here  

### 📌 Modern AI = Rational Agents

## 3️⃣ Weak AI vs Strong AI

### 🔹 Weak AI (Narrow AI) ✅ (Current world)

- Solves specific tasks  
- No consciousness  

Examples:
- Google Assistant  
- Face recognition  
- Recommendation systems  

### 🔹 Strong AI (General AI) ❌ (Not achieved)

- Human-level intelligence  
- Can reason in any domain  
- Self-aware

## 5️⃣ Simple AI Code Example (Rule-Based AI)

This is AI but NOT ML.

### 🔹 Problem:

If temperature is high → turn AC ON

### 🔹 Basic Code Example (Python)


```python
def ac_controller(temperature):
    if temperature > 30:
        return "AC ON"
    else:
        return "AC OFF"

print(ac_controller(35))
```

📌 This is:

AI ✔️

Logical reasoning ✔️

No training ✔️

No dataset ✔️

### 6️⃣ Advanced AI Code (Rule-Based Expert System)

```python
def medical_expert_system(symptoms):
    if "fever" in symptoms and "cough" in symptoms:
        return "Possible Flu"
    elif "fever" in symptoms and "rash" in symptoms:
        return "Possible Dengue"
    else:
        return "Consult a doctor"


print(medical_expert_system(["fever", "cough"]))
```

## 📌 Used in:

- Expert Systems  
- Early AI applications  
- Medical diagnosis tools
