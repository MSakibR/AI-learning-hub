# 🔹 Backward Chaining (Rule-Based Systems)

## 1️⃣ BASIC THEORY (Easy & Intuitive)

Backward Chaining is a goal-driven reasoning method in AI.

Start with a goal → check which rules can satisfy it → recursively check their preconditions → continue until facts are found.

Think of it as “Can I prove the goal from what I know?”

Works from top-down.

### Simple Real-Life Example

**Goal:** Diagnose Flu

**Rule:** `IF Fever AND Cough THEN Flu`

**Facts:**

- Fever = True
- Cough = True

**Backward Chaining Steps:**

1. Is Flu true? → Check rules → need Fever AND Cough  
2. Check facts → Fever = True, Cough = True → goal satisfied

## 2️⃣ ADVANCED THEORY (Exam / Practical Level)

### 🔸 Steps of Backward Chaining

1. Start with the goal  
2. Find rules where goal = conclusion  
3. Check if preconditions are satisfied:  
   - If yes → goal satisfied  
   - If not → treat preconditions as sub-goals  
4. Repeat recursively until:  
   - All preconditions are true → goal satisfied  
   - Or no applicable rule → fail

### 🔹 Comparison with Forward Chaining

| Feature        | Forward Chaining       | Backward Chaining  |
|----------------|---------------------|------------------|
| Approach       | Data-driven           | Goal-driven       |
| Direction      | Bottom-up             | Top-down          |
| Starts With    | Facts                 | Goal              |
| Used For       | Finding all conclusions | Checking specific goal |
| Efficiency     | Can compute all facts | Efficient for specific goals |

### 🔹 Formula / Logic Representation

**Goal:**  
$$
Q
$$

**Rule:**  
$$
P_1 \wedge P_2 \rightarrow Q
$$

**Backward Chaining:**  
To prove $$Q$$, prove $$P_1, P_2$$

## 3️⃣ BASIC CODE EXAMPLE (Python)


````python
# Simple backward chaining
facts = {"fever", "cough"}
rules = [
    ({"fever", "cough"}, "flu"),
    ({"flu", "fatigue"}, "rest")
]

def backward_chaining(goal, facts, rules):
    if goal in facts:
        return True
    for conditions, conclusion in rules:
        if conclusion == goal:
            if all(backward_chaining(cond, facts, rules) for cond in conditions):
                return True
    return False

goal = "flu"
print(f"Is '{goal}' true? ->", backward_chaining(goal, facts, rules))
````

### Output
`Is 'flu' true? -> True`

## 4️⃣ ADVANCED CODE EXAMPLE (Recursive Goal-Driven Inference)

````python
rules = [
    ({"hair"}, "mammal"),
    ({"mammal", "gives_milk"}, "cow"),
    ({"feathers"}, "bird"),
    ({"bird", "flies"}, "sparrow"),
]

facts = {"hair", "gives_milk", "feathers"}


def backward_chaining(goal, facts, rules, indent=0):
    print("  " * indent + f"Trying to prove: {goal}")
    if goal in facts:
        print("  " * indent + f"Found in facts: {goal}")
        return True

    for conditions, conclusion in rules:
        if conclusion == goal:
            if all(
                backward_chaining(cond, facts, rules, indent + 1) for cond in conditions
            ):
                print("  " * indent + f"Rule applied: {conditions} -> {goal}")
                return True
    print("  " * indent + f"Cannot prove: {goal}")
    return False


goal = "cow"
print("Goal Satisfied?", backward_chaining(goal, facts, rules))
````

### Output
```python
Trying to prove: cow
  Trying to prove: mammal
  Found in facts: hair
  Trying to prove: gives_milk
  Found in facts: gives_milk
Rule applied: {'mammal', 'gives_milk'} -> cow
Goal Satisfied? True
```

## 5️⃣ ADVANTAGES

- ✅ Efficient for specific goals
- ✅ Saves computation if goal is limited
- ✅ Intuitive for diagnosis systems

## 6️⃣ LIMITATIONS

- ❌ Not efficient if many goals are needed
- ❌ Can be recursive-heavy
- ❌ Requires complete and correct rule base

## 7️⃣ IMPORTANT NOTES

- Backward Chaining = Goal-Driven
- Starts from the goal → works backward
- Often used in:  
  - Expert systems  
  - Diagnostics (medical, mechanical)  
- Complements Forward Chaining
