# Rule-Based Systems (Artificial Intelligence)

## 1️⃣ BASIC THEORY (Easy & Intuitive)

### What is a Rule-Based System?

A Rule-Based System (RBS) is an AI system that:  

- Uses IF–THEN rules  
- Makes decisions by matching facts with rules  

👉 Think of it like human decision-making rules:

IF traffic light is red  
THEN stop the car  

### Core Idea

`IF condition(s) are true`\
`THEN take action / conclude something`

### Why Rule-Based Systems?

- Simple  
- Transparent (easy to explain)  
- Good for expert systems  

### Real-life Examples

- Medical diagnosis systems  
- Loan approval systems  
- Spam filters (basic ones)  
- University rule engines (grading rules)  

## 2️⃣ BASIC COMPONENTS OF RULE-BASED SYSTEM

### 1. Knowledge Base

Contains facts + rules  

**Example:**  

`Fact: fever = true`\
`Rule: IF fever AND cough THEN flu `

### 2. Inference Engine

- Applies rules to facts  
- Decides which rule to fire  

### 3. Working Memory

- Stores current facts  

### 4. Explanation Facility (optional)

- Explains why a decision was made  

## 3️⃣ BASIC WORKING (Step-by-Step)

- Read facts  
- Match facts with rules  
- Fire applicable rules  
- Add new facts  
- Repeat until no rule applies  

## 4️⃣ BASIC CODE EXAMPLE (Python)

### Simple Rule-Based Decision System


````python
def rule_based_system(facts):
    if facts.get("fever") and facts.get("cough"):
        return "Diagnosis: Flu"
    elif facts.get("fever"):
        return "Diagnosis: Viral Fever"
    else:
        return "Diagnosis: Unknown"

# Facts
facts = {
    "fever": True,
    "cough": True
}

print(rule_based_system(facts))
````

### Output
```python
Diagnosis: Flu
```

## 5️⃣ ADVANCED THEORY (Exam + Practical Level)

### Rule Representation

Rules are usually represented as:

`IF P1 ∧ P2 ∧ P3`\
`THEN Q`


Where:  

- P1, P2, P3 = conditions  
- Q = conclusion  

### Types of Rule Execution

- Forward Chaining (Data-Driven)  
- Backward Chaining (Goal-Driven)  

We already touched this earlier — here it fits naturally.  

## 6️⃣ ADVANCED RULE-BASED SYSTEM (Forward Chaining)

### Example: Animal Classification


````python
rules = [
    ({"hair"}, "mammal"),
    ({"mammal", "gives_milk"}, "cow"),
    ({"feathers"}, "bird"),
]

facts = {"hair", "gives_milk"}


def forward_chaining(rules, facts):
    inferred = set()

    while True:
        applied = False
        for conditions, conclusion in rules:
            if conditions.issubset(facts) and conclusion not in facts:
                facts.add(conclusion)
                inferred.add(conclusion)
                applied = True
        if not applied:
            break

    return inferred


print(forward_chaining(rules, facts))
````

### Output
```python
{'mammal', 'cow'}
```

## 7️⃣ FORMULA / LOGICAL VIEW (If Needed)

In propositional logic form:

**(hair ∧ gives_milk) → cow**

Inference uses Modus Ponens:

P → Q\
P\
∴ Q

## 8️⃣ ADVANTAGES & LIMITATIONS

### ✅ Advantages

- Easy to understand  
- Easy to modify  
- Transparent reasoning  

### ❌ Limitations

- Poor scalability  
- No learning ability  
- Hard to manage large rule sets  
- Rule conflicts possible  

## 9️⃣ IMPORTANT EXAM NOTES ⚠️

- Rule-Based Systems are knowledge-driven  
- Used heavily in Expert Systems  
- Inference Engine is the core  
- Rules ≠ Machine Learning (no training)
