# 🔹 Forward Chaining (Rule-Based Systems)

## 1️⃣ BASIC THEORY (Easy & Intuitive)

Forward Chaining is a data-driven reasoning method in AI.

Start with known facts → apply rules → deduce new facts → repeat until goal reached.

Think of it as “What can I conclude from what I know?”

Works from bottom-up.

### Simple Real-Life Example

**Facts:**

- Fever = True
- Cough = True

**Rule:**

`IF Fever AND Cough THEN Flu`

**Process (Forward Chaining):**

1. Check facts → Fever AND Cough are true
2. Fire rule → conclude Flu
3. Add Flu to facts → maybe more rules fire

## 2️⃣ ADVANCED THEORY (Exam / Practical Level)

### 🔸 Components

- **Fact Base (Working Memory):** Stores all known facts
- **Rule Base:** Contains IF-THEN rules
- **Inference Engine:** Applies rules on facts

### 🔸 Algorithm (Step-by-Step)

1. Initialize fact base  
2. Loop:  
   - Find all rules whose conditions are satisfied by facts  
   - Fire one or more rules  
   - Add conclusions to facts  
3. Repeat until no more rules can fire or goal reached

### 🔹 Formula / Logic Representation

**Rule:**  
$$
P_1 \wedge P_2 \wedge \dots \wedge P_n \rightarrow Q
$$

**Facts:**  
$$
P_1, P_2, \dots, P_n
$$

**Inference (Modus Ponens):**  
$$
Q
$$

Forward Chaining uses Modus Ponens iteratively.

## 3️⃣ BASIC CODE EXAMPLE (Python)


````python
# Simple forward chaining example
facts = {"fever", "cough"}

rules = [
    ({"fever", "cough"}, "flu"),
    ({"flu", "fatigue"}, "rest")
]

def forward_chaining(facts, rules):
    inferred = set()
    while True:
        applied = False
        for conditions, conclusion in rules:
            if conditions.issubset(facts) and conclusion not in facts:
                facts.add(conclusion)
                inferred.add(conclusion)
                applied = True
        if not applied:
            break
    return inferred

print("Inferred facts:", forward_chaining(facts, rules))
````

### Output
```python
Inferred facts: {'flu'}
```

Only flu is inferred because fatigue was not in facts.

## 4️⃣ ADVANCED CODE EXAMPLE (Goal-Oriented Forward Chaining)
Example: Animal Classification

````python
rules = [
    ({"hair"}, "mammal"),
    ({"mammal", "gives_milk"}, "cow"),
    ({"feathers"}, "bird"),
    ({"bird", "flies"}, "sparrow"),
]

facts = {"hair", "gives_milk", "feathers"}


def forward_chaining(rules, facts):
    inferred = set()
    while True:
        applied = False
        for conds, concl in rules:
            if conds.issubset(facts) and concl not in facts:
                facts.add(concl)
                inferred.add(concl)
                print(f"Rule fired: {conds} -> {concl}")
                applied = True
        if not applied:
            break
    return inferred


print("Inferred Facts:", forward_chaining(rules, facts))
````

### Output
```python
Rule fired: {'hair'} -> mammal
Rule fired: {'mammal', 'gives_milk'} -> cow
Rule fired: {'feathers'} -> bird
Inferred Facts: {'mammal', 'cow', 'bird'}
```

## 5️⃣ ADVANTAGES

- ✅ Simple, intuitive, transparent
- ✅ Works well with data-driven problems
- ✅ Easy to implement in small expert systems

## 6️⃣ LIMITATIONS

- ❌ Can be inefficient if many rules exist
- ❌ Might fire unnecessary rules
- ❌ Works only for forward reasoning (goal-driven may need backward chaining)

## 7️⃣ IMPORTANT NOTES

- Forward Chaining = Data-Driven
- Starts from known facts → deduces new facts
- Often used in diagnosis systems (medical, technical)
- Inference is done via Modus Ponens repeatedly
