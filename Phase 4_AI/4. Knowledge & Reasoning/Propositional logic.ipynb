# Propositional Logic (PL)

## 🔹 1. BASIC THEORY

Propositional Logic is the simplest form of logic used in AI for reasoning.  

Also called Boolean Logic / Sentential Logic  

Deals with propositions (statements that are either True or False)  

### Example:

- “It is raining.” → Proposition P  
- “I will take an umbrella.” → Proposition Q  

### Propositions can be connected using logical operators:

| Operator       | Symbol | Meaning           |
|----------------|--------|-----------------|
| AND            | ∧      | Both true        |
| OR             | ∨      | At least one true|
| NOT            | ¬      | Negation         |
| IMPLIES        | →      | If…then          |
| BICONDITIONAL  | ↔      | If and only if   |

### 🔹 Simple Example

“If it rains, I will take an umbrella.”

$$
P \rightarrow Q
$$

Where:  
- P = It rains  
- Q = I take umbrella  

## 🔹 2. ADVANCED THEORY

- **Knowledge in PL:** A set of propositions and logical connectives  
- **Inference in PL:** Using rules to derive new facts  

### Important Rules

- **Modus Ponens:**  
$$
(P \rightarrow Q), P \vdash Q
$$  
If P implies Q, and P is true → Q is true  

- **Modus Tollens:**  
$$
(P \rightarrow Q), \neg Q \vdash \neg P
$$  

- **Disjunctive Syllogism:**  
$$
(P \vee Q), \neg P \vdash Q
$$  

- **Conjunction / Simplification:**  
From P ∧ Q → P or Q individually  

- **Resolution Rule:** Used in AI for automated reasoning  

### 🔹 FORMAL REPRESENTATION

- **Syntax:** How to write propositions  
- **Semantics:** What they mean (True/False)  
- **Knowledge Base (KB):** Collection of propositional statements  

### 📌 Example KB:

1. P → Q  
2. Q → R  
3. P  

**Query:** Is R true?  
**Answer:** Yes, using Chaining:  

- P → Q, P is true → Q is true  
- Q → R, Q is true → R is true ✅  

## 🔹 3. BASIC CODE EXAMPLE (Python)


````python
# Propositional Logic using simple rules
def modus_ponens(P_implies_Q, P):
    if P_implies_Q and P:
        return True
    return False


P = True  # It is raining
Q = False  # Will take umbrella (initially)
P_implies_Q = True

Q = modus_ponens(P_implies_Q, P)
print("Will take umbrella?", Q)
````

### 📌 Output:
```python
Will take umbrella? True
```


## 🔹 4. ADVANCED CODE EXAMPLE (Forward Chaining in PL)

````python
# Knowledge Base
KB = {
    "P": True,  # It is raining
    "P->Q": True,  # If it rains then I will take umbrella
    "Q->R": True,  # If I take umbrella then I stay dry
}


def forward_chaining(KB):
    inferred = {}

    # Apply rules
    if KB.get("P") and KB.get("P->Q"):
        inferred["Q"] = True
    if inferred.get("Q") and KB.get("Q->R"):
        inferred["R"] = True

    return inferred


print("Inferred facts:", forward_chaining(KB))
````

### 📌 Output:
```python
Inferred facts: {'Q': True, 'R': True}
```

## 🔹 5. Important Notes

- PL is limited: Cannot represent objects, relations, or quantities

- FOL (First-Order Logic) solves this limitation

- PL is foundation for Expert Systems
