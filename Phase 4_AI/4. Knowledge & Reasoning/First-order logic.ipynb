# First-Order Logic (FOL)

## 🔹 1. BASIC THEORY

First-Order Logic (FOL) extends Propositional Logic by allowing:

- **Objects** → things in the world  
- **Predicates** → properties or relations of objects  
- **Variables** → general statements about objects  
- **Quantifiers** → “all” or “some”  

### In simple words:

FOL = Propositional Logic + objects + relations + quantifiers  

### 🔹 Why FOL?

- PL is limited: only true/false propositions  
- FOL can express complex statements  

**Example:**  

All humans are mortal.  
Socrates is a human.  
→ Socrates is mortal  

In FOL:

$$
Human(Socrates) \wedge \forall x (Human(x) \rightarrow Mortal(x)) \Rightarrow Mortal(Socrates)
$$

## 🔹 Components

| Component   | Example                        |
|------------|--------------------------------|
| Constants  | Socrates, Plato                |
| Variables  | x, y                           |
| Predicates | Human(x), Mortal(x)            |
| Functions  | Father(x), Mother(x)           |
| Quantifiers| ∀x (for all), ∃x (exists)     |
| Connectives| ∧, ∨, ¬, →, ↔                 |

## 🔹 2. ADVANCED THEORY

### 🔸 Quantifiers

- **Universal (∀):** “For all x”  
  Example: $$\forall x (Human(x) \rightarrow Mortal(x))$$  

- **Existential (∃):** “There exists x”  
  Example: $$\exists x (Human(x) \wedge Philosopher(x))$$  

### 🔸 Functions & Predicates

- **Predicate:** Property or relation (returns True/False)  
- **Function:** Maps objects → objects  
  Example: Father(Socrates) = Plato  

### 🔸 Inference in FOL

- Forward Chaining → Deduce facts from rules  
- Backward Chaining → Check goal using rules  
- Unification → Match variables to constants or other variables  

## 🔹 3. FORMAL REPRESENTATION

**FOL Statement Example:**

$$
\forall x (Cat(x) \rightarrow Animal(x)) \\
Cat(Tom) \\
\Rightarrow Animal(Tom)
$$

- Universal quantifier: ∀  
- Predicate: Cat(x), Animal(x)  
- Inference: Deductive reasoning  

## 🔹 4. BASIC CODE EXAMPLE (FOL Representation)


````python
# Simple FOL example using Python dict
KB = {
    "Human": ["Socrates", "Plato"],
    "Mortal": []
}

def infer_mortal(KB):
    for person in KB["Human"]:
        KB["Mortal"].append(person)
    return KB["Mortal"]

print("Mortals:", infer_mortal(KB))
````

### 📌 Output:
```python
Mortals: ['Socrates', 'Plato']
```

## 🔹 5. ADVANCED CODE EXAMPLE (Forward Chaining with FOL)

````python
# Represent facts and rules
facts = {"Human": ["Socrates", "Plato"], "Mortal": []}
rules = [("Human", "Mortal")]  # If Human(x) -> Mortal(x)

def forward_chaining_FOL(facts, rules):
    inferred = set()
    for rule in rules:
        premise, conclusion = rule
        for obj in facts.get(premise, []):
            if obj not in facts.get(conclusion, []):
                facts[conclusion].append(obj)
                inferred.add(obj)
    return inferred

print("Inferred Mortals:", forward_chaining_FOL(facts, rules))
````

📌 Output:
```python
Inferred Mortals: {'Socrates', 'Plato'}
```
