# Unification in AI

## 🔹 1. BASIC THEORY

Unification is a key concept in AI reasoning (especially in FOL).

**Unification = Finding a substitution that makes two logical expressions identical**

### In simple words:

When reasoning with variables, we often need to match patterns.  
Unification tells us how to substitute variables so two expressions become equal.

### 🔹 Simple Example (Intuitive)

- Expression 1: Father(x) = John  
- Expression 2: Father(Socrates) = y  

**Unification finds substitutions:**

- x → Socrates  
- y → John  

After substitution: both expressions match.

### 🔹 Why Unification?

- Essential for Automated Theorem Proving  
- Used in Forward / Backward Chaining  
- Used in Prolog and logic-based AI systems  

## 🔹 2. ADVANCED THEORY

### 🔸 Algorithmic Idea

- Compare two expressions:  
  - If both are constants: must be equal  
  - If one is variable: substitute variable with other expression  
  - If both are functions: their names must match, unify arguments recursively  
- If no conflicts, expressions unify  
- Else, unification fails  

### 🔸 Terminology

- **Substitution (θ):** Mapping variable → term  
- **Most General Unifier (MGU):** Simplest substitution that unifies expressions  
- **Conflict:** When constants differ → unification fails  

### 🔹 Example

Expressions:  

$$
P(x, f(y)) \text{ and } P(g(z), f(a))
$$

**Unification Steps:**

- Compare outer function → both P → ok  
- Arguments: x vs g(z) → substitute x → g(z)  
- Next: f(y) vs f(a) → substitute y → a  

**MGU:**  

$$
\{ x \rightarrow g(z),\ y \rightarrow a \}
$$

## 🔹 3. BASIC CODE EXAMPLE (Python Simplified)


````python
# Simple unification example
def unify(expr1, expr2, subs={}):
    if expr1 == expr2:
        return subs
    elif isinstance(expr1, str) and expr1.islower():  # variable
        subs[expr1] = expr2
        return subs
    elif isinstance(expr2, str) and expr2.islower():  # variable
        subs[expr2] = expr1
        return subs
    return None

expr1 = "x"
expr2 = "John"

print("Substitution:", unify(expr1, expr2))
````

📌 Output:
```python
Substitution: {'x': 'John'}
```

## 🔹 4. ADVANCED CODE EXAMPLE (Recursive FOL Unification)

````python
def unify_recursive(x, y, subs={}):
    if subs is None:
        return None
    elif x == y:
        return subs
    elif isinstance(x, str) and x.islower():  # variable
        subs[x] = y
        return subs
    elif isinstance(y, str) and y.islower():
        subs[y] = x
        return subs
    elif isinstance(x, list) and isinstance(y, list):
        if len(x) != len(y):
            return None
        for xi, yi in zip(x, y):
            subs = unify_recursive(xi, yi, subs)
        return subs
    else:
        return None


# Example
expr1 = ["P", "x", ["f", "y"]]
expr2 = ["P", "g(z)", ["f", "a"]]

print("MGU:", unify_recursive(expr1, expr2))
````

📌 Output:
```python
MGU: {'x': 'g(z)', 'y': 'a'}
```

## 🔹 5. Important Notes

- Most General Unifier (MGU) ensures flexible, minimal substitutions
