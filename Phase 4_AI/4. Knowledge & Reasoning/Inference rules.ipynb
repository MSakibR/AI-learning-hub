# Inference Rules in AI

## 🔹 1. BASIC THEORY

Inference Rules are the logic rules that allow agents to deduce new facts from existing knowledge.  

### In simple words:

“If the facts and rules are known, how can we logically conclude something new?”  

### 🔹 Why Important?

- Core of reasoning in AI  
- Used in:  
  - Expert Systems  
  - Prolog and logic programming  
  - Forward / Backward Chaining  

**Example:**  

- Fact: “It is raining” → Raining  
- Rule: “If it is raining, the ground is wet” → Raining → WetGround  
- **Inference:** WetGround ✅  

## 🔹 2. BASIC TYPES OF INFERENCE RULES

- **Modus Ponens (Most Used)**  

$$
P \rightarrow Q,\, P \vdash Q
$$  

If P implies Q and P is true → Q is true  

- **Modus Tollens**  

$$
P \rightarrow Q,\, \neg Q \vdash \neg P
$$  

If P implies Q and Q is false → P is false  

- **Hypothetical Syllogism**  

$$
(P \rightarrow Q) \wedge (Q \rightarrow R) \vdash P \rightarrow R
$$  

- **Conjunction**  

$$
P, Q \vdash P \wedge Q
$$  

- **Simplification**  

$$
P \wedge Q \vdash P
$$  

- **Disjunctive Syllogism**  

$$
P \vee Q, \neg P \vdash Q
$$  

- **Resolution (Advanced, Used in AI)**  
  - Combines clauses to deduce new facts  
  - Central to Automated Theorem Proving  

## 🔹 3. FORMULAS / REPRESENTATION

- **Forward Chaining:**  
  Start with facts, apply inference rules → deduce new facts  

- **Backward Chaining:**  
  Start with goal, check if any rule/fact supports it → recursively verify  

- **Resolution Example:**  

$$
\neg P \vee Q,\, P \vdash Q
$$  

- Convert to CNF (Conjunctive Normal Form)  
- Combine clauses to infer new facts  

## 🔹 4. BASIC CODE EXAMPLE (Forward Chaining, PL)


````python
# Knowledge Base
facts = {"Raining": True}
rules = [("Raining", "WetGround")]

def forward_inference(facts, rules):
    inferred = set()
    for premise, conclusion in rules:
        if facts.get(premise):
            inferred.add(conclusion)
    return inferred

print("Inferred Facts:", forward_inference(facts, rules))
````

📌 Output:
```python
Inferred Facts: {'WetGround'}
```

## 🔹 5. ADVANCED CODE EXAMPLE (Chaining + FOL)

````python
# Facts & Rules
facts = {"Human": ["Socrates"], "Mortal": []}
rules = [("Human", "Mortal")]  # Human(x) -> Mortal(x)


def forward_chaining_FOL(facts, rules):
    inferred = set()
    for premise, conclusion in rules:
        for obj in facts.get(premise, []):
            if obj not in facts.get(conclusion, []):
                facts[conclusion].append(obj)
                inferred.add(obj)
    return inferred


print("Inferred Mortals:", forward_chaining_FOL(facts, rules))
````

📌 Output:
```python
Inferred Mortals: {'Socrates'}
```

🔹 6. Important Notes

- Inference Rules + Unification = Core of reasoning in AI

- Forward Chaining: Data-driven, starts with facts → goal

- Backward Chaining: Goal-driven, starts with goal → verify facts

- Resolution Rule: Powerful for automated theorem proving
