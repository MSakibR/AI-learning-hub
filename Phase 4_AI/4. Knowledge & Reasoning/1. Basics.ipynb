# Knowledge & Reasoning — BASICS

## 🔹 1. Basic Theory (Beginner-friendly)

Knowledge & Reasoning (K&R) is a core part of AI that allows an agent to:

- Represent knowledge about the world  
- Reason about it to make decisions  

### In simple words:

“Give the agent knowledge about the world, so it can think and act intelligently.”

### 🔹 Why Knowledge & Reasoning?

- AI needs more than just data  
- ML/DL can learn patterns, but classical AI uses explicit knowledge and logic  

**Example:**  
“All humans are mortal. Socrates is a human. → Socrates is mortal.” ✅

This is reasoning from knowledge

## 🔹 Components of Knowledge & Reasoning

| Component               | Explanation                                      |
|------------------------|-------------------------------------------------|
| Knowledge Base (KB)     | Stores facts & rules about the world           |
| Inference Engine        | Draws new conclusions from KB                  |
| Language / Representation | How facts/rules are written (logic, frames, semantic networks) |

## 🔹 Types of Knowledge

- **Declarative Knowledge ✅**  
  - Facts & rules  
  - Example: “Socrates is a human”  

- **Procedural Knowledge ❌ (used less in classical AI)**  
  - How to do tasks  
  - Example: “How to drive a car”  

### 📌 Classical AI mainly uses declarative knowledge + reasoning.

## 🔹 Basic Reasoning Types

- **Deductive Reasoning**  
  - General → Specific  
  - Example: All humans are mortal → Socrates is human → Socrates is mortal  

- **Inductive Reasoning**  
  - Specific → General  
  - Example: Sun rises every day → Sun rises tomorrow  

- **Abductive Reasoning**  
  - Best explanation  
  - Example: Grass is wet → It probably rained  

## 🔹 Basic Code Example (Simple Deductive Reasoning)


```python
# Knowledge Base
facts = ["Socrates is a human", "All humans are mortal"]

# Reasoning
def deduce(facts):
    if "Socrates is a human" in facts and "All humans are mortal" in facts:
        return "Socrates is mortal"
    return "Cannot deduce"

print(deduce(facts))
```

### 📌 Output:

```python
Socrates is mortal
```

## 🔹 Advanced Theory (Interview / Exam Level)

### Knowledge Representation Languages

- Propositional Logic (PL)  
- First-Order Logic (FOL)  
- Semantic Networks, Frames, Ontologies  

### Inference Types

- **Forward Chaining:** From facts → derive new facts  
- **Backward Chaining:** From goal → check facts  

### Unification

- Matching symbols/variables in logical statements  

### Inference Rules

- **Modus Ponens:** “If P → Q, P is true → Q is true”  
- Resolution, Conjunction, Disjunction  

## 🔹 Important Notes

✔ Classical AI is logic-heavy  
✔ Knowledge & Reasoning = Agent thinks like a human using facts and rules
