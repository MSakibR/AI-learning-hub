# Intelligent Agents

## 🔹 1. BASIC THEORY (Easy Understanding)

An Intelligent Agent is anything that can perceive its environment and act upon it to achieve a goal.

### 📌 Simple definition:

Agent = Perceive + Decide + Act

### Examples:

- A human (eyes → brain → hands)  
- A robot (sensors → controller → motors)  
- A software agent (input → logic → output)  

## 🔹 Agent Components

| Component     | Meaning                                   |
|--------------|--------------------------------------------|
| Sensors      | Collect information from environment       |
| Actuators    | Perform actions                            |
| Environment  | World where agent operates                 |
| Agent Function | Maps perception → action                |

### 📌 Example:

Camera → Image → Decision → Steering

## 🔹 2. FORMAL (ADVANCED) THEORY

### 🔸 Formal Definition

An intelligent agent is an autonomous entity that perceives its environment through sensors and acts upon that environment through actuators to maximize its performance measure.

### 🔸 Agent Function

Mathematically:

$$
f : P^* \rightarrow A
$$

Where:

- $P^*$ = sequence of perceptions  
- $A$ = actions



### 📌 Meaning:

Based on everything the agent has perceived so far, it chooses an action.

## 🔹 3. PEAS Framework (VERY IMPORTANT)

PEAS = Performance, Environment, Actuators, Sensors

Used to describe agent tasks clearly.

### Example: Self-driving car

| Element      | Description                          |
|-------------|--------------------------------------|
| Performance | Safety, speed, fuel efficiency       |
| Environment | Roads, traffic, pedestrians          |
| Actuators   | Steering, brakes, accelerator        |
| Sensors     | Cameras, LiDAR, GPS                  |

### 📌 This will be asked in exams & interviews.

## 🔹 4. BASIC CODE EXAMPLE (Simple Agent)

### 🔸 Goal:

Turn on light if room is dark


```python
def simple_agent(light_intensity):
    if light_intensity < 40:
        return "Light ON"
    else:
        return "Light OFF"

print(simple_agent(20))
```

✔ Perception → light_intensity
✔ Action → Light ON/OFF
✔ Intelligence → Rule-based decision

### 🔹 5. ADVANCED CODE EXAMPLE (Agent with Environment)

```python
class Environment:
    def __init__(self, temperature):
        self.temperature = temperature


class IntelligentAgent:
    def act(self, env):
        if env.temperature > 30:
            return "Turn AC ON"
        else:
            return "Turn AC OFF"


env = Environment(35)
agent = IntelligentAgent()

print(agent.act(env))
```

## 📌 This shows:

- Separation of agent and environment  
- Real AI architecture style  

## 🔹 6. Rational Agent (KEY CONCEPT)

A rational agent:

- Does the right action  

Based on:

- Percept sequence  
- Knowledge of environment  
- Available actions  
- Performance measure  

### 📌 Rational ≠ Perfect
### 📌 Rational = Best decision with available info
