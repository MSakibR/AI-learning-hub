# PEAS DESCRIPTION

## 🔹 1. BASIC THEORY

PEAS is a framework to define an AI agent’s problem clearly.

PEAS stands for:

| Letter | Meaning             | Description                       |
|--------|-------------------|-----------------------------------|
| P      | Performance Measure| How well the agent is doing       |
| E      | Environment        | Where the agent operates          |
| A      | Actuators          | What the agent can do             |
| S      | Sensors            | How the agent perceives the environment |

### 📌 Purpose:

- Helps design rational agents  
- Makes requirements explicit and measurable  

## 🔹 2. ADVANCED THEORY

- **Performance measure:** Should capture all aspects of “success”  
  Example: Safety, speed, efficiency  

- **Environment types:**  
  - Fully observable / partially observable  
  - Deterministic / stochastic  
  - Static / dynamic  
  - Discrete / continuous  
  - Single-agent / multi-agent  

- **Actuators:** Hardware/software outputs  
- **Sensors:** Cameras, LiDAR, keyboard input, APIs  

### Example (Self-driving car):

| Element      | Description                          |
|-------------|--------------------------------------|
| Performance | Reach destination safely, minimize fuel |
| Environment | Roads, traffic, pedestrians, weather |
| Actuators   | Steering, brakes, accelerator        |
| Sensors     | GPS, cameras, LiDAR, speedometer     |

## 🔹 3. CODE EXAMPLE (Python Representation of PEAS)


```python
class PEASAgent:
    def __init__(self, performance, environment, actuators, sensors):
        self.performance = performance
        self.environment = environment
        self.actuators = actuators
        self.sensors = sensors

# Example for self-driving car
car_agent = PEASAgent(
    performance = ["Safety", "Efficiency"],
    environment = ["Roads", "Traffic", "Weather"],
    actuators = ["Steering", "Brakes", "Accelerator"],
    sensors = ["Camera", "LiDAR", "GPS"]
)

print(vars(car_agent))
```

### 📌 Output:

```python
{
    "performance": ["Safety", "Efficiency"],
    "environment": ["Roads", "Traffic", "Weather"],
    "actuators": ["Steering", "Brakes", "Accelerator"],
    "sensors": ["Camera", "LiDAR", "GPS"],
}
```

## 🔹 4. ADVANCED NOTES

- PEAS helps in exam questions: “Define the performance measure, actuators, sensors of X agent”

- For every AI agent, you can describe using PEAS
