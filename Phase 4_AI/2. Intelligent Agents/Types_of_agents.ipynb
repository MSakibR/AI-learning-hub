# Types of Intelligent Agents

Intelligent agents can be categorized based on how they perceive the environment, reason, and act.

## 🔹 1. BASIC THEORY

There are mainly 5 types of agents:

- **Simple Reflex Agents**  
  - Act only on current percept  
  - No memory of past states  
  - Rule-based (if condition → action)  

- **Model-Based Reflex Agents**  
  - Maintain an internal state  
  - Uses current percept + past info  
  - Can handle partially observable environments  

- **Goal-Based Agents**  
  - Make decisions to achieve a specific goal  
  - Uses reasoning to select actions  

- **Utility-Based Agents**  
  - Measures happiness/success via a utility function  
  - Chooses actions that maximize utility  

- **Learning Agents**  
  - Improves performance over time  
  - Learns from experience  

## 🔹 2. ADVANCED THEORY (Interview/Exam Level)

| Agent Type       | Strength                     | Weakness                       | Example                       |
|-----------------|------------------------------|--------------------------------|-------------------------------|
| Simple Reflex    | Fast, easy                   | Cannot handle new situations   | Traffic light controller       |
| Model-Based      | Handles partial observability| Needs model maintenance        | Robot vacuum                   |
| Goal-Based       | Flexible, smart              | Planning overhead              | Chess AI                       |
| Utility-Based    | Optimal, trade-off handling  | Needs well-defined utility     | Self-driving cars              |
| Learning         | Adapts to environment        | Needs experience & data        | AlphaGo, recommendation systems|

### 🔹 Key Concept:

Most real-world AI agents are hybrids: Model-based + Goal-based + Learning

## 🔹 3. FORMULAS (If needed)

Utility-based agent selects the action $a$ that maximizes expected utility:

$$
a = \arg\max_{a \in \text{Actions}} EU(a)
$$

Where $EU(a)$ = Expected Utility of action $a$

## 🔹 4. BASIC CODE EXAMPLES

### 🔸 1. Simple Reflex Agent


```python
def reflex_agent(percept):
    if percept == "Hungry":
        return "Eat"
    elif percept == "Thirsty":
        return "Drink"
    else:
        return "Rest"

print(reflex_agent("Hungry"))
```

### 🔸 2. Model-Based Reflex Agent

```python
class ModelBasedAgent:
    def __init__(self):
        self.state = "Healthy"

    def perceive(self, percept):
        self.state = percept  # update internal model
        if self.state == "Hungry":
            return "Eat"
        else:
            return "Rest"


agent = ModelBasedAgent()
print(agent.perceive("Hungry"))
```

### 🔸 3. Goal-Based Agent

```python
class GoalAgent:
    def __init__(self, goal):
        self.goal = goal

    def act(self, state):
        if state != self.goal:
            return f"Take action to reach {self.goal}"
        else:
            return "Goal achieved!"


agent = GoalAgent("Reach Home")
print(agent.act("At Office"))
```

### 🔸 4. Utility-Based Agent

```python
def utility_agent(options):
    utilities = {"Walk": 5, "Drive": 10, "Taxi": 7}
    best_action = max(utilities, key=utilities.get)
    return best_action


print(utility_agent(None))
```

### 🔸 5. Learning Agent (Basic Example)

```python
class LearningAgent:
    def __init__(self):
        self.knowledge = {}

    def learn(self, percept, action):
        self.knowledge[percept] = action

    def act(self, percept):
        return self.knowledge.get(percept, "Explore")


agent = LearningAgent()
agent.learn("Hungry", "Eat")
print(agent.act("Hungry"))
```

### 🔹 5. ADVANCED CODE EXAMPLE (Hybrid Agent)

```python
class HybridAgent:
    def __init__(self, goal):
        self.goal = goal
        self.state = None
        self.experience = {}

    def perceive(self, percept):
        self.state = percept

    def learn(self, percept, action):
        self.experience[percept] = action

    def act(self):
        if self.state in self.experience:
            return self.experience[self.state]
        elif self.state != self.goal:
            return f"Take action to reach {self.goal}"
        else:
            return "Goal achieved!"


agent = HybridAgent("Home")
agent.perceive("Office")
agent.learn("Office", "Drive")
print(agent.act())
```

## 📌 This shows:

- Model-based (keeps state)  
- Goal-based (has goal)  
- Learning (adapts to experience)  

## 🔹 6. Important Notes

✔ Most real-world AI systems are hybrid agents  
✔ Reflex agents are rare in complex systems  
✔ Utility-based agents are used in self-driving cars and game AI  
✔ Learning agents are core to modern AI (ML + DL)
