We will now start Search Algorithms.

## 🔹 1. BASIC THEORY

Search algorithms are used by agents to find a sequence of actions to reach a goal from the initial state.

### Key terms:

- **State:** Representation of the environment  
- **Initial State:** Where search starts  
- **Goal State:** Target state  
- **Actions:** Moves from one state to another  
- **Path/Plan:** Sequence of actions to reach goal  

### Types of search:

- **Uninformed (Blind) Search** – No knowledge about goal location  
  - BFS, DFS, Uniform Cost  

- **Informed (Heuristic) Search** – Uses knowledge to guide search  
  - Greedy Best-First, A*  

## 🔹 2. FORMULAS / Concepts

- **Cost function:** $g(n)$ = path cost from initial state to node $n$  
- **Heuristic function:** $h(n)$ = estimated cost from node $n$ to goal  

- **Evaluation function (A*):**

$$
f(n) = g(n) + h(n)
$$

- **Greedy uses:**

$$
f(n) = h(n)
$$

## 🔹 3. ALGORITHMS BASICS

| Algorithm          | Type       | Key Idea                    | Completeness | Optimality                  |
|------------------|------------|----------------------------|--------------|----------------------------|
| BFS               | Uninformed | Explore level by level      | Yes          | Yes (if cost=1)            |
| DFS               | Uninformed | Go deep first               | No           | No                         |
| Uniform Cost      | Uninformed | Expand lowest path-cost first | Yes        | Yes                        |
| Greedy Best-First | Informed   | Expand lowest $h(n)$       | No           | No                         |
| A*                | Informed   | Expand lowest $f(n)=g+h$   | Yes          | Yes (if $h$ admissible)    |

## 🔹 4. CODE EXAMPLES (Python)

### 🔸 BFS (Uninformed)


```python
from collections import deque

def bfs(graph, start, goal):
    queue = deque([ [start] ])
    visited = set()
    
    while queue:
        path = queue.popleft()
        node = path[-1]
        if node == goal:
            return path
        if node not in visited:
            visited.add(node)
            for neighbor in graph.get(node, []):
                new_path = path + [neighbor]
                queue.append(new_path)

graph = {
    'A': ['B','C'],
    'B': ['D','E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("BFS Path:", bfs(graph, 'A', 'F'))
```

### 🔸 DFS (Uninformed)

```python
def dfs(graph, start, goal, path=[], visited=set()):
    path = path + [start]
    visited.add(start)
    if start == goal:
        return path
    for neighbor in graph.get(start, []):
        if neighbor not in visited:
            res = dfs(graph, neighbor, goal, path, visited)
            if res:
                return res
    return None


print("DFS Path:", dfs(graph, "A", "F"))
```

### 🔸 Uniform Cost Search

```python
import heapq


def ucs(graph, start, goal):
    queue = [(0, start, [start])]
    visited = set()

    while queue:
        cost, node, path = heapq.heappop(queue)
        if node == goal:
            return path, cost
        if node not in visited:
            visited.add(node)
            for neighbor, c in graph.get(node, []):
                heapq.heappush(queue, (cost + c, neighbor, path + [neighbor]))


graph = {
    "A": [("B", 1), ("C", 4)],
    "B": [("D", 2), ("E", 5)],
    "C": [("F", 1)],
    "D": [],
    "E": [("F", 1)],
    "F": [],
}

print("UCS Path and cost:", ucs(graph, "A", "F"))
```

### 🔸 Greedy Best-First Search

```python
def greedy_search(graph, heuristics, start, goal):
    from queue import PriorityQueue

    pq = PriorityQueue()
    pq.put((heuristics[start], [start]))
    visited = set()

    while not pq.empty():
        _, path = pq.get()
        node = path[-1]
        if node == goal:
            return path
        if node not in visited:
            visited.add(node)
            for neighbor in graph.get(node, []):
                new_path = path + [neighbor]
                pq.put((heuristics[neighbor], new_path))


heuristics = {"A": 6, "B": 4, "C": 2, "D": 4, "E": 2, "F": 0}
print("Greedy Path:", greedy_search(graph, heuristics, "A", "F"))
```

### 🔸 A* Search

```python
def a_star(graph, heuristics, start, goal):
    from queue import PriorityQueue

    pq = PriorityQueue()
    pq.put((heuristics[start], 0, [start]))
    visited = set()

    while not pq.empty():
        f, g, path = pq.get()
        node = path[-1]
        if node == goal:
            return path, g
        if node not in visited:
            visited.add(node)
            for neighbor, cost in graph.get(node, []):
                g_new = g + cost
                f_new = g_new + heuristics[neighbor]
                pq.put((f_new, g_new, path + [neighbor]))


print("A* Path and cost:", a_star(graph, heuristics, "A", "F"))
```

## 🔹 5. Advanced Notes

- BFS: Good for shortest path in unweighted graphs

- DFS: Memory efficient but may fail on infinite graphs

- UCS: Optimal for variable costs

- Greedy: Fast but not always optimal

- A*: Most used in AI pathfinding, needs admissible heuristics
