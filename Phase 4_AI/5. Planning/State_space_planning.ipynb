# PLANNING — STATE-SPACE PLANNING

## 🔹 1. BASIC THEORY (Easy Understanding)

State-Space Planning treats planning as a search problem.

The planner searches through different states of the world to find a path from the initial state to the goal state.

### 🔹 Key Idea

- **State** = representation of the world at a moment  
- **Action** = transforms one state into another  
- **Plan** = sequence of actions (path)  

📌 Very similar to Search Algorithms you already learned (BFS, DFS, A*).

### 🔹 Simple Example

**Initial State:**  
At(Home)

**Goal State:**  
At(University)

**Actions:**

- Walk(Home → BusStop)  
- RideBus(BusStop → University)  

Planning = finding this action sequence

## 🔹 2. ADVANCED THEORY (Exam / Interview Level)

### 🔸 Two Types of State-Space Planning

#### 1️⃣ Forward State-Space Planning (Progression)

- Start from initial state  
- Apply actions  
- Move forward until goal is reached  

#### 2️⃣ Backward State-Space Planning (Regression)

- Start from goal  
- Work backwards  
- Find actions that could achieve the goal  

📌 Mapping:

- Forward planning ≈ Forward Chaining  
- Backward planning ≈ Backward Chaining  

### 🔹 State-Space Planning Components

| Component       | Description                               |
|-----------------|-------------------------------------------|
| Initial State   | Starting point                            |
| Goal State      | Desired condition                         |
| Actions         | Operators with preconditions & effects    |
| State Space     | All possible states                       |
| Search Strategy | BFS, DFS, A*, UCS                         |

## 🔹 3. FORMAL MODEL

An action $A$ is defined as:

- $Pre(A)$ → Preconditions  
- $Add(A)$ → Positive effects  
- $Del(A)$ → Negative effects  

Action is applicable if:

$$
Pre(A) \subseteq CurrentState
$$

After applying action:

$$
NewState = (CurrentState - Del(A)) \cup Add(A)
$$

## 🔹 4. BASIC CODE EXAMPLE (Forward State-Space Planning)

### 🔸 Problem:

Move from A → D


````python
from collections import deque


def state_space_planning(initial, goal, transitions):
    queue = deque([(initial, [initial])])
    visited = set()

    while queue:
        state, path = queue.popleft()
        if state == goal:
            return path
        if state not in visited:
            visited.add(state)
            for next_state in transitions.get(state, []):
                queue.append((next_state, path + [next_state]))


states = {"A": ["B", "C"], "B": ["D"], "C": ["D"], "D": []}

print("Plan:", state_space_planning("A", "D", states))
````

📌 Output:
```python
Plan: ['A', 'B', 'D']
```

## 🔹 5. ADVANCED CODE EXAMPLE (Action-Based State-Space Planning)

````python
class Action:
    def __init__(self, name, pre, add, delete):
        self.name = name
        self.pre = set(pre)
        self.add = set(add)
        self.delete = set(delete)


def apply_action(state, action):
    if action.pre.issubset(state):
        return (state - action.delete) | action.add
    return None


# Initial state
initial_state = {"AtHome"}

# Goal
goal_state = {"AtUniversity"}

# Actions
actions = [
    Action("GoToBusStop", {"AtHome"}, {"AtBusStop"}, {"AtHome"}),
    Action("TakeBus", {"AtBusStop"}, {"AtUniversity"}, {"AtBusStop"}),
]

# Forward planning
frontier = [(initial_state, [])]
visited = set()

while frontier:
    state, plan = frontier.pop(0)
    if goal_state.issubset(state):
        print("Plan Found:", plan)
        break
    for action in actions:
        new_state = apply_action(state, action)
        if new_state and frozenset(new_state) not in visited:
            visited.add(frozenset(new_state))
            frontier.append((new_state, plan + [action.name]))
````

📌 Output:
```python
Plan Found: ['GoToBusStop', 'TakeBus']
```

## 🔹 6. COMPARISON: GOAL STACK vs STATE-SPACE PLANNING

| Feature        | Goal Stack        | State-Space            |
|---------------|------------------|------------------------|
| Approach      | Goal-driven      | Search-driven          |
| Direction     | Top-down         | Forward / Backward     |
| Optimal       | ❌ No            | ✅ With BFS / A*       |
| Complexity    | Simple           | High                   |
| Used in       | Early AI         | Modern planners        |
