# PLANNING — GOAL STACK PLANNING

## 🔹 1. BASIC THEORY (Easy + Intuitive)

Planning in AI means:

Finding a sequence of actions that transforms the initial state into a goal state.

### 🔹 What is Goal Stack Planning?

Goal Stack Planning is a classical AI planning technique where:

- Goals are stored in a stack  
- The agent:  
  - Takes the top goal  
  - Tries to satisfy it  
  - If not possible → pushes sub-goals  
  - Continues until all goals are achieved  

📌 Think of it like solving problems top-down.

### 🔹 Simple Real-Life Example

**Goal:**  
👉 Have breakfast  

**Sub-goals:**

- Have food  
- Food must be cooked  
- Ingredients must be available  

📌 These goals are solved one by one using a stack.

## 🔹 2. ADVANCED THEORY (Exam / Interview Level)

### 🔸 Key Concepts

| Term                | Meaning                          |
|---------------------|----------------------------------|
| Initial State        | Starting condition               |
| Goal State           | Desired condition                |
| Operators (Actions)  | Actions that change state        |
| Preconditions        | Conditions needed before an action |
| Effects              | Result after action              |
| Goal Stack           | Stack holding goals & actions    |

### 🔹 Working Mechanism

- Push final goal onto stack  
- If top of stack is:  
  - **Goal** → check if satisfied  
    - If yes → pop  
    - If no → push an action that achieves it  
  - **Action** →  
    - Push its preconditions as new goals  
    - Execute action once preconditions satisfied  
- Repeat until stack is empty  

📌 This is a means–ends analysis approach.

## 🔹 3. FORMAL REPRESENTATION (Conceptual)

Let:

- $G$ = Goal  
- $A$ = Action  
- $Pre(A)$ = Preconditions  
- $Eff(A)$ = Effects  

Action is applicable if:

$$
Pre(A) \subseteq CurrentState
$$

After execution:

$$
CurrentState = (CurrentState - DeleteList) \cup AddList
$$

## 🔹 4. BASIC CODE EXAMPLE (Simplified Goal Stack Planner)

### 🔸 Problem:

Goal → Room is clean


````python
# Initial state
state = {"room_dirty", "tools_available"}

# Goal stack
goal_stack = ["room_clean"]

# Actions
actions = {
    "clean_room": {
        "preconditions": {"tools_available"},
        "effects": {"room_clean"}
    }
}

while goal_stack:
    goal = goal_stack.pop()
    
    if goal in state:
        continue
    
    if goal == "room_clean":
        action = "clean_room"
        goal_stack.append(action)
        goal_stack.extend(actions[action]["preconditions"])
    
    elif goal == "clean_room":
        state.add("room_clean")
        print("Action executed: Clean room")

print("Final State:", state)
````

📌 Output:
```python
Action executed: Clean room
Final State: {'room_dirty', 'tools_available', 'room_clean'}
```

## 🔹 5. ADVANCED CODE EXAMPLE (Structured Goal Stack Planning)

````python
class GoalStackPlanner:
    def __init__(self, initial_state):
        self.state = set(initial_state)
        self.stack = []
    
    def plan(self, goal, actions):
        self.stack.append(goal)
        
        while self.stack:
            top = self.stack.pop()
            
            if top in self.state:
                continue
            
            if top in actions:
                # It's an action
                self.state |= actions[top]["effects"]
                print(f"Execute: {top}")
            else:
                # It's a goal
                for action, details in actions.items():
                    if top in details["effects"]:
                        self.stack.append(action)
                        self.stack.extend(details["preconditions"])
                        break

# Define actions
actions = {
    "clean_room": {
        "preconditions": {"tools_available"},
        "effects": {"room_clean"}
    }
}

planner = GoalStackPlanner({"tools_available"})
planner.plan("room_clean", actions)

print("Final State:", planner.state)
````

📌 Output:
```python
Execute: clean_room
Final State: {'tools_available', 'room_clean'}
```

## 🔹 6. IMPORTANT NOTES (You SHOULD Know)

✔ Goal Stack Planning is not optimal  
✔ Can fail in complex or cyclic environments  
✔ Works well for simple deterministic problems  
✔ Foundation of early AI planners (STRIPS-like systems)
