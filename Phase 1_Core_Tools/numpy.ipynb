# Numpy Basic

## What is NumPy?

. NumPy = Numerical Python

. Efficient arrays, matrices, and mathematical operations.

. Faster than regular Python lists for calculations.

1. Importing NumPy

```python
import numpy as np
```

2. Creating Arrays

```python
a = np.array([1, 2, 3, 4, 5])
print(a)
```

```python
b = np.array([[1, 2, 3], [4, 5, 6]])
print(b)
```

```python
zeros = np.zeros((3, 3))
ones = np.ones((2, 4))
print(zeros)
print(ones)
```

```python
c = np.arange(0, 10, 2)  # 0 2 4 6 8
print(c)
```

3. Array Operations

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print(x + y)  # [5 7 9]
print(x * y)  # [4 10 18]
print(x - y)  # [-3 -3 -3]
print(x / y)  # [0.25 0.4 0.5]
```

4. Array Functions

```python
a = np.array([1, 2, 3, 4, 5])
print(np.mean(a))  # Average
print(np.sum(a))  # Sum
print(np.max(a))  # Max value
print(np.min(a))  # Min value
print(np.std(a))  # Standard deviation
```

5. Indexing & Slicing

```python
a = np.array([10, 20, 30, 40, 50])
print(a[0])  # 10
print(a[1:4])  # [20 30 40]
print(a[-1])  # 50
```

### Mini Project: Temperature Conversion

```python
temps_c = np.array([0, 10, 20, 30, 40])
temps_f = temps_c * 9 / 5 + 32
print(temps_f)  # [ 32.  50.  68.  86. 104.]
```

```python
# task 1 - Create a NumPy array from 1 to 20 and print all even numbers.

import numpy as np

arr = np.arange(1, 21)  # Array from 1 to 20
even_numbers = arr[arr % 2 == 0]  # Select even numbers

print("Even numbers:", even_numbers)
```

```python
# task 2 - Create a 3×3 array of zeros and ones (combine them using np.vstack or np.hstack).

import numpy as np

row1 = np.zeros(3)
row2 = np.ones(3)
row3 = np.zeros(3)

arr = np.vstack((row1, row2, row3))
print(arr)
```

```python
# task 3  Find mean, sum, max, min, and standard deviation of [5, 10, 15, 20, 25].

import numpy as np

arr = np.array([5, 10, 15, 20, 25])

print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))
print("Std:", np.std(arr))
```

```python
# task 4 - Slice the array [10,20,30,40,50,60,70] to get [20,30,40].

import numpy as np

arr = np.array([10, 20, 30, 40, 50, 60, 70])

slice_arr = arr[1:4]
print(slice_arr)
```

```python
# task 5 - Convert array [0, 5, 10, 15, 20] from Celsius to Fahrenheit.

import numpy as np

celsius = np.array([0, 5, 10, 15, 20])
fahrenheit = celsius * 9 / 5 + 32

print("Fahrenheit:", fahrenheit)
```
