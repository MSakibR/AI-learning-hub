7. Filtering Data

```python
import pandas as pd

df = pd.read_csv("students.csv")
df[df["Age"] > 22]
df[df["CGPA"] >= 3.8]
```

8. Handling Missing Values

```python
import pandas as pd

df = pd.read_csv("students.csv")
df.isnull().sum()  # check nulls
df.fillna(0)  # replace nulls
df.dropna()  # remove rows with nulls
```

9. Adding New Columns

```python
import pandas as pd

df = pd.read_csv("students.csv")
df["Pass"] = df["CGPA"] > 3.5
```

### 🔥 Mini Project: Student Performance Analyzer

```python
import pandas as pd

df = pd.read_csv("students.csv")

print("Top Students:")
print(df[df["CGPA"] > 3.75])

df["Grade"] = df["CGPA"].apply(lambda x: "A+" if x >= 3.75 else "A")
print(df)
```

```python
# task 1

import pandas as pd

data = {
    "Name": ["Sakib", "Rahim", "Karim", "Nadim"],
    "Department": ["CSE", "EEE", "CSE", "BBA"],
    "CGPA": [3.80, 3.50, 3.90, 3.20],
    "Credits": [120, 110, 130, 100],
}

df = pd.DataFrame(data)
print(df)
```

```python
# task 2
import pandas as pd

df = pd.DataFrame({"Marks": [85, 92, 76, 88, 95], 
                   "Attendance": [90, 80, 85, 70, 95]
                   })

# Average marks
avg_marks = df["Marks"].mean()

# Maximum attendance
max_attendance = df["Attendance"].max()

# Rows where marks > 90
high_marks = df[df["Marks"] > 90]

print("Average Marks:", avg_marks)
print("Maximum Attendance:", max_attendance)
print("Marks > 90:")
print(high_marks)
```

```python
# task 3
import pandas as pd

data = {
    "Name": ["Sakib", "Hasan", "Nusrat"],
    "Age": [22, 23, 21],
    "City": ["Dhaka", "Chittagong", "Rajshahi"],
}

df = pd.DataFrame(data)
df.to_csv("data.csv", index=False)
```

```python
# task 4
import pandas as pd

df = pd.DataFrame({"ID": [1, 2, 3, 4], 
                   "Score": [50, None, 80, None]})

df["Score"] = df["Score"].fillna(0)
print(df)
```

```python
# Task 5
import pandas as pd

df = pd.DataFrame({"ID": [1, 2, 3, 4], 
                   "Score": [50, None, 80, None]})

df["Status"] = df["Score"].apply(lambda x: "Pass" if x >= 60 else "Fail")

print(df)
```
