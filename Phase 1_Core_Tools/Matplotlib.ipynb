# Matplotlib (Data Visualization)

## 🎯 Why Matplotlib?

Because you will visualize:

distributions
trends
correlations
ML results
confusion matrices
loss curves (DL)

Without Matplotlib, ML/DL becomes blind.

1. Import Matplotlib

```python
import matplotlib.pyplot as plt
```

2. Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 40, 50]

plt.plot(x, y)
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

3. Bar Chart

```python
import matplotlib.pyplot as plt

subjects = ["CSE", "EEE", "ME", "BBA"]
students = [120, 90, 60, 70]

plt.bar(subjects, students)
plt.title("Department-wise Students")
plt.show()
```

4. Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 9, 11]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

5. Histogram

```python
import matplotlib.pyplot as plt

marks = [55, 60, 65, 70, 72, 75, 80, 85, 90, 92, 95]

plt.hist(marks, bins=5)
plt.title("Marks Distribution")
plt.show()
```

6. Pie Chart

```python
import matplotlib.pyplot as plt

sizes = [40, 30, 20, 10]
labels = ["A+", "A", "A-", "Others"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("CGPA Distribution")
plt.show()
```

BONUS — Multiple Plots Together

```python
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Line Plot")

plt.subplot(1, 2, 2)
plt.bar(["A", "B", "C"], [5, 7, 3])
plt.title("Bar Chart")

plt.show()
```
