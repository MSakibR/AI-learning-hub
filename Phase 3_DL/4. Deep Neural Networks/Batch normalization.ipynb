# 📌 Batch Normalization

---

## 🧠 1️⃣ What is Batch Normalization?

**Batch Normalization (BatchNorm)** is a technique to **normalize the input of each layer** so that:

- Mean ≈ 0  
- Standard deviation ≈ 1

It reduces **internal covariate shift**, meaning neurons get inputs of **similar distribution every epoch**.

---

### ✅ Benefits

- Speeds up training  
- Allows **higher learning rates**  
- Reduces dependency on careful initialization  
- Often improves **final accuracy**

---

### 🔹 Mathematical View

For each mini-batch  

$$
B = \{ x_1, x_2, \dots, x_m \}
$$

Compute **mean** and **variance**:

$$
\mu_B = \frac{1}{m} \sum_{i=1}^{m} x_i
$$

$$
\sigma_B^2 = \frac{1}{m} \sum_{i=1}^{m} (x_i - \mu_B)^2
$$

Normalize each input:

$$
\hat{x}_i = \frac{x_i - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}}
$$

Scale & shift (trainable parameters):

$$
y_i = \gamma \hat{x}_i + \beta
$$

Where:

- $\gamma, \beta$ = learned during training  
- $\epsilon$ = small constant for numerical stability  

---

## 🟢 2️⃣ Why BatchNorm Works

- Helps **gradient flow** in deep networks → reduces vanishing/exploding gradients  
- Works well with **ReLU** and **Dropout**  
- Makes **training faster**

---

## 🟢 3️⃣ BatchNorm Example in Keras


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout

model = Sequential([
    Dense(128, activation='relu', input_shape=(30,)),
    BatchNormalization(),
    Dropout(0.5),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

✅ Notes:

Place BatchNorm after Dense and before activation (Keras automatically handles ordering)

Combine BatchNorm + Dropout → better generalization

## 🔵 4️⃣ BatchNorm in PyTorch

```python
import torch
import torch.nn as nn
import torch.optim as optim

class DNNWithBN(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, 128)
        self.bn1 = nn.BatchNorm1d(128)
        self.dropout1 = nn.Dropout(0.5)
        self.fc2 = nn.Linear(128, 64)
        self.bn2 = nn.BatchNorm1d(64)
        self.dropout2 = nn.Dropout(0.3)
        self.fc3 = nn.Linear(64, 1)
    
    def forward(self, x):
        x = torch.relu(self.bn1(self.fc1(x)))
        x = self.dropout1(x)
        x = torch.relu(self.bn2(self.fc2(x)))
        x = self.dropout2(x)
        x = torch.sigmoid(self.fc3(x))
        return x

# Example usage
model = DNNWithBN(input_dim=30)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

## ✅ Key Points

- `BatchNorm1d` for fully connected layers  
- `BatchNorm2d` for convolutional layers  
- Trainable parameters `gamma` and `beta` are updated automatically


## 🟣 5️⃣ Advanced Tips — BatchNorm

- Use **after Dense** + **before activation**  
- Combine with Dropout carefully (**Dropout after BatchNorm**)  
- Can increase **learning rate safely**  
- Often eliminates need for **careful weight initialization**

## 🔹 6️⃣ DNN Summary (Dense + Dropout + BatchNorm)


```python
Input → Dense → BatchNorm → Activation → Dropout → Dense → BatchNorm → Activation → Dropout → Output
```
