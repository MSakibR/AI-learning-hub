## 1️⃣ Neural Network Basics

A **Neural Network (NN)** is a computational model that mimics the way the human brain works.  
It consists of layers of neurons that transform input data into output predictions.

---

## 🧠 Structure of a Neural Network

### 🔹 Neurons

Neurons are the **basic computational units** of a neural network.

Each neuron:
- Receives inputs
- Applies weights
- Adds a bias
- Passes the result through an activation function

### 🔢 Mathematical Representation

\[
y = f(w_1x_1 + w_2x_2 + \dots + w_nx_n + b)
\]

Where:
- \(x_i\) = input  
- \(w_i\) = weight  
- \(b\) = bias  
- \(f\) = activation function  
- \(y\) = output  

---

### 🔹 Layers

- **Input Layer:** Receives raw features from the dataset  
- **Hidden Layer(s):** Perform intermediate computations  
- **Output Layer:** Produces the final prediction  

---

### 🔹 Weights & Biases

- **Weights** determine the importance of each input  
- **Bias** allows the model to shift the activation function  

---

## ➡️ Forward Propagation

Forward propagation is the process of passing input data through the network to obtain an output.

Each neuron computes:

\[
\text{output} = \text{activation}(\text{weighted sum} + \text{bias})
\]

or equivalently,

\[
y = f(wx + b)
\]

---

## 📉 Loss Function

The **loss function** measures how far the network’s prediction is from the actual value.

### Examples:
- **Mean Squared Error (MSE)** → used for regression
- **Cross-Entropy Loss** → used for classification

---

## 🔁 Backward Propagation

Backward propagation:
- Adjusts weights and biases based on the error
- Minimizes the loss function
- Uses **Gradient Descent** for optimization

---

## 📌 Key Terms to Remember

- Neuron / Node  
- Layer (Input, Hidden, Output)  
- Weights & Bias  
- Activation Function  
- Forward Propagation  
- Loss Function  
- Backward Propagation  


## 🧩 Perceptron

The **Perceptron** is the simplest type of neural network, invented by **Frank Rosenblatt (1958)**.  
It is basically a **single neuron model** used to classify **linearly separable data**.

---

## 1️⃣ Perceptron Structure

| Component | Description |
|--------|------------|
| **Inputs** \(x_1, x_2, \dots, x_n\) | Features of the data |
| **Weights** \(w_1, w_2, \dots, w_n\) | Importance of each input |
| **Bias** \(b\) | Shifts the activation function |
| **Weighted Sum** | \(z = w_1x_1 + w_2x_2 + \dots + w_nx_n + b\) |
| **Activation Function** | Produces output based on weighted sum |

---

## 2️⃣ Perceptron Formula

The perceptron computes the output as:

\[
y =
\begin{cases}
1, & \text{if } \sum_{i=1}^{n} w_i x_i + b > 0 \\
0, & \text{otherwise}
\end{cases}
\]

---

### 🔹 Where:

- \(x_i\) = input features  
- \(w_i\) = weights  
- \(b\) = bias  
- \(y\) = output (0 or 1)

---

## ⚙️ Activation Function

The perceptron uses the **Step Activation Function**:

- Outputs **1** if the weighted sum is positive
- Outputs **0** otherwise

This makes the perceptron suitable for **binary classification problems**.

---

## 📌 Key Notes

- Can only solve **linearly separable problems**
- Cannot solve problems like **XOR**
- Foundation of modern neural networks


### 3️⃣ Python Implementation of a Perceptron

```python
import numpy as np

# Step activation function
def step_function(x):
    return 1 if x > 0 else 0

# Perceptron class
class Perceptron:
    def __init__(self, input_size, lr=0.1, epochs=10):
        self.weights = np.zeros(input_size)
        self.bias = 0
        self.lr = lr
        self.epochs = epochs

    # Training the perceptron
    def fit(self, X, y):
        for _ in range(self.epochs):
            for i in range(len(X)):
                linear_output = np.dot(X[i], self.weights) + self.bias
                y_pred = step_function(linear_output)
                # Update rule
                update = self.lr * (y[i] - y_pred)
                self.weights += update * X[i]
                self.bias += update

    # Predict new data
    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return np.array([step_function(i) for i in linear_output])

# Example dataset: AND logic gate
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([0, 0, 0, 1])

# Create Perceptron and train
p = Perceptron(input_size=2)
p.fit(X, y)

# Testing predictions
print("Weights:", p.weights)
print("Bias:", p.bias)
print("Predictions:", p.predict(X))
```

## 4️⃣ Explanation of the Code

| Step | Code | Description |
|----|------|-------------|
| Activation | `step_function(x)` | Returns **1** if input > 0, else **0** |
| Initialization | `self.weights = np.zeros(input_size)` | Start weights as **0** |
| Training loop | `for _ in range(self.epochs)` | Repeat training for all epochs |
| Linear output | `linear_output = np.dot(X[i], self.weights) + self.bias` | Computes weighted sum + bias |
| Prediction | `y_pred = step_function(linear_output)` | Applies activation function |
| Update rule | `update = lr * (y[i] - y_pred)` | Adjusts weights based on prediction error |
| Weight update | `self.weights += update * X[i]` | Modifies weights |
| Bias update | `self.bias += update` | Modifies bias |
| Prediction on new data | `predict()` | Computes outputs for test data |

---

## ✅ Key Note

This Perceptron can learn **basic logic gates** such as:
- AND
- OR
- NOT

⚠️ However, it **cannot solve non-linear problems** like **XOR**.  
To handle such problems, we need **Multi-Layer Neural Networks (MLP)**.


```python
import numpy as np

# Step activation function
def step_function(x):
    return 1 if x >= 0 else 0

# Perceptron
class Perceptron:
    def __init__(self, input_size, lr=0.1):
        self.weights = np.zeros(input_size)
        self.bias = 0
        self.lr = lr

    def predict(self, x):
        linear_output = np.dot(x, self.weights) + self.bias
        return step_function(linear_output)

    def train(self, X, y, epochs=10):
        for _ in range(epochs):
            for xi, yi in zip(X, y):
                pred = self.predict(xi)
                error = yi - pred
                self.weights += self.lr * error * xi
                self.bias += self.lr * error

# Dataset: AND gate
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

p = Perceptron(input_size=2)
p.train(X, y)

# Predictions
for xi in X:
    print(f"Input: {xi}, Predicted: {p.predict(xi)}")
```

```python
from sklearn.linear_model import Perceptron

X = [[0,0],[0,1],[1,0],[1,1]]
y = [0,0,0,1]

clf = Perceptron(max_iter=1000, eta0=0.1)
clf.fit(X, y)

print("Predictions:", clf.predict(X))
```
