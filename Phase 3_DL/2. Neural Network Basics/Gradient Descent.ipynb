## 📉 Gradient Descent (GD)

**Gradient Descent** is an optimization algorithm used to minimize the **loss function** by updating the model’s **weights and biases**.

---

## 1️⃣ Intuition

Imagine you are standing on a **hill** and want to reach the **lowest point (valley)**.

- Gradient Descent tells you **which direction to move**
- It also tells you **how big a step to take**
- The step size is controlled by the **learning rate** \(\alpha\)

---

## 2️⃣ Gradient Descent Formula

For a weight \(w\) and loss function \(L(w)\):

\[
w := w - \alpha \frac{\partial L}{\partial w}
\]

### 🔹 Where:
- \(w\) = weight  
- \(\alpha\) = learning rate  
- \(\frac{\partial L}{\partial w}\) = derivative of loss w.r.t. weight  

---

### 🔹 Bias Update Formula

Similarly, for bias \(b\):

\[
b := b - \alpha \frac{\partial L}{\partial b}
\]

This process is **repeated iteratively** until the loss is minimized.

---

## 3️⃣ Types of Gradient Descent

| Type | Description |
|---|---|
| **Batch Gradient Descent** | Uses all training samples to compute the gradient. Slow for large datasets. |
| **Stochastic Gradient Descent (SGD)** | Uses one sample at a time. Faster but noisy updates. |
| **Mini-batch Gradient Descent** | Uses small batches of data. Combines speed and stability. |

---

## 📌 Key Notes

- Too large \(\alpha\) → may overshoot the minimum  
- Too small \(\alpha\) → slow convergence  
- Gradient Descent is the backbone of **neural network training**


## 4️⃣ Python Example

Let’s see a simple linear regression example using Gradient Descent:

```python
import numpy as np

# Sample dataset
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])  # y = 2*x

# Initialize parameters
w = 0
b = 0
lr = 0.01  # learning rate
epochs = 1000
n = len(X)

# Gradient Descent loop
for _ in range(epochs):
    # Predicted values
    y_pred = w * X + b
    
    # Compute gradients
    dw = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)
    
    # Update parameters
    w = w - lr * dw
    b = b - lr * db

print("Trained weight:", w)
print("Trained bias:", b)
```

```python
import numpy as np

# Dataset
X = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Initialize weight and bias
w, b = 0.0, 0.0
lr = 0.01
epochs = 1000

# Training loop
for _ in range(epochs):
    y_pred = w*X + b
    error = y_pred - y
    dw = (2/len(X)) * np.dot(error, X)
    db = (2/len(X)) * np.sum(error)
    w -= lr * dw
    b -= lr * db

print(f"Trained weight: {w}, bias: {b}")
```

## 5️⃣ Explanation

| Step | Code | Description |
|---|---|---|
| Prediction | `y_pred = w * X + b` | Computes current predictions |
| Gradient of weight | `dw = (-2/n) * np.sum(X * (y - y_pred))` | Determines how much to change the weight |
| Gradient of bias | `db = (-2/n) * np.sum(y - y_pred)` | Determines how much to change the bias |
| Update | `w = w - lr * dw` | Adjusts weight in the direction of reducing loss |
| Epochs | `for _ in range(epochs)` | Repeats updates until convergence |

---

## ✅ Takeaway

**Gradient Descent** is the backbone of **neural network training**.  
Once we understand Gradient Descent, we can move on to **Activation Functions**, which allow neural networks to learn **non-linear patterns**.


```python
import numpy as np

# Dataset: Simple linear relationship
X = np.array([1, 2, 3, 4, 5])
y = np.array([3, 6, 9, 12, 15])  # y = 3*x

# Initialize parameters
w = 0
b = 0
lr = 0.01  # learning rate
epochs = 1000
n = len(X)

for i in range(epochs):
    # Forward pass: prediction
    y_pred = w * X + b
    
    # Compute gradients
    dw = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)
    
    # Update weights
    w = w - lr * dw
    b = b - lr * db
    
    # Print loss every 200 epochs
    if i % 200 == 0:
        loss = np.mean((y - y_pred)**2)
        print(f"Epoch {i}, Loss: {loss}")

print("Trained weight:", w)
print("Trained bias:", b)
```

```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Dataset: OR gate
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([0, 1, 1, 1])

# Initialize weights and bias
w = np.zeros(X.shape[1])
b = 0
lr = 0.1
epochs = 10000

for _ in range(epochs):
    # Forward pass
    linear_output = np.dot(X, w) + b
    y_pred = sigmoid(linear_output)
    
    # Compute gradients (derivative of loss for MSE)
    dw = np.dot(X.T, (y_pred - y)) / X.shape[0]
    db = np.sum(y_pred - y) / X.shape[0]
    
    # Update weights
    w -= lr * dw
    b -= lr * db

print("Trained weights:", w)
print("Trained bias:", b)
print("Predictions:", np.round(sigmoid(np.dot(X, w) + b)))
```
