## ⚡ Activation Functions

In neural networks, the **activation function** decides whether a neuron should be activated or not.  
It introduces **non-linearity** into the network — without activation functions, a neural network would behave like a **linear model**, no matter how many layers you add.

---

## 1️⃣ Why Activation Functions?

- Without activation functions, the network can only model **linear relationships**
- With activation functions, the network can model **complex, non-linear relationships**
- Helps proper **gradient flow during backpropagation**

---

## 2️⃣ Common Activation Functions

| Function | Formula | Range | Python Example |
|--------|--------|-------|----------------|
| **Step / Threshold** | \( f(x) = \begin{cases} 1, & x > 0 \\ 0, & \text{otherwise} \end{cases} \) | 0 or 1 | `lambda x: 1 if x > 0 else 0` |
| **Sigmoid** | \( f(x) = \frac{1}{1 + e^{-x}} \) | 0 to 1 | `1 / (1 + np.exp(-x))` |
| **Tanh** | \( f(x) = \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} \) | -1 to 1 | `np.tanh(x)` |
| **ReLU** | \( f(x) = \max(0, x) \) | 0 to ∞ | `np.maximum(0, x)` |
| **Leaky ReLU** | \( f(x) = \begin{cases} x, & x > 0 \\ 0.01x, & \text{otherwise} \end{cases} \) | -∞ to ∞ | `np.where(x > 0, x, 0.01 * x)` |

---

## 📌 Key Notes

- **Sigmoid** → good for probability outputs, but suffers from vanishing gradients  
- **Tanh** → zero-centered, better than sigmoid  
- **ReLU** → most commonly used, fast and effective  
- **Leaky ReLU** → fixes the “dying ReLU” problem  

---

## 3️⃣ Graphs of Activation Functions

We can visualize them in Python:


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

# Functions
sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)
relu = np.maximum(0, x)
leaky_relu = np.where(x>0, x, 0.01*x)

# Plot
plt.figure(figsize=(10,6))
plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, tanh, label="Tanh")
plt.plot(x, relu, label="ReLU")
plt.plot(x, leaky_relu, label="Leaky ReLU")
plt.title("Activation Functions")
plt.legend()
plt.grid(True)
plt.show()
```

## 4️⃣ Python Examples Using Activation Functions
Sigmoid Example

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([-2, -1, 0, 1, 2])
print("Sigmoid:", sigmoid(x))
```

```python
# Sigmoid: [0.119, 0.268, 0.5, 0.731, 0.881]
```

### Tanh Example

```python
def tanh(x):
    return np.tanh(x)

x = np.array([-2, -1, 0, 1, 2])
print("Tanh:", tanh(x))
```

```python
# Tanh: [-0.964, -0.762, 0, 0.762, 0.964]
```

### ReLU Example

```python
def relu(x):
    return np.maximum(0, x)

x = np.array([-2, -1, 0, 1, 2])
print("ReLU:", relu(x))
```

```python
# ReLU: [0, 0, 0, 1, 2]
```

### Leaky ReLU Example

```python
def leaky_relu(x):
    return np.where(x>0, x, 0.01*x)

x = np.array([-2, -1, 0, 1, 2])
print("Leaky ReLU:", leaky_relu(x))
```

```python
# Leaky ReLU: [-0.02, -0.01, 0, 1, 2]
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)

sigmoid = 1/(1+np.exp(-x))
tanh = np.tanh(x)
relu = np.maximum(0,x)
leaky_relu = np.where(x>0, x, 0.01*x)

plt.plot(x, sigmoid, label='Sigmoid')
plt.plot(x, tanh, label='Tanh')
plt.plot(x, relu, label='ReLU')
plt.plot(x, leaky_relu, label='Leaky ReLU')
plt.legend()
plt.show()
```

  <h3>✅ Key Takeaways</h3>
  <ul>
    <li><strong>Sigmoid:</strong> Good for binary classification, smooth gradient, but can saturate.</li>
    <li><strong>Tanh:</strong> Zero-centered, better than sigmoid for hidden layers.</li>
    <li><strong>ReLU:</strong> Most popular for hidden layers in deep networks. Fast and avoids vanishing gradient.</li>
    <li><strong>Leaky ReLU:</strong> Solves ReLU’s “dead neuron” problem.</li>
  </ul>
