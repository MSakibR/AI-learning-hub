
### 1️⃣ Forward Propagation

Forward propagation calculates the **output** of the network from the **input**, using **weights, bias, and activation functions**.

$$
a^{[l]} = f \Big( W^{[l]} a^{[l-1]} + b^{[l]} \Big)
$$

Where:  
- $a^{[l]}$ = activation (output) of layer $l$  
- $W^{[l]}$ = weight matrix of layer $l$  
- $a^{[l-1]}$ = activation from previous layer $(l-1)$  
- $b^{[l]}$ = bias vector of layer $l$  
- $f(\cdot)$ = activation function


---

### 2️⃣ Backward Propagation

Backward propagation **updates the weights and biases** by calculating gradients using the **chain rule**, aiming to **minimize the loss**:

$$
W := W - \eta \frac{\partial L}{\partial W}, \quad
b := b - \eta \frac{\partial L}{\partial b}
$$

Where:  
- $W$ = weights  
- $b$ = biases  
- $\eta$ = learning rate  
- $L$ = loss function  
- $\frac{\partial L}{\partial W}, \frac{\partial L}{\partial b}$ = gradients of the loss with respect to weights and biases


---




### Python Example: Simple 1-layer NN forward/backward propagation

```python
# Forward
def forward(x, w, b):
    z = np.dot(x, w) + b
    a = 1/(1+np.exp(-z)) # Sigmoid activation
    return a

# Backward (gradient)
def backward(x, y, a, w, b, lr):
    dz = a - y
    dw = np.dot(x.T, dz)
    db = np.sum(dz)
    w -= lr * dw
    b -= lr * db
    return w, b

# Example data
X = np.array([[0],[1]])
y = np.array([[0],[1]])
w = np.array([[0.5]])
b = 0
lr = 0.1

for i in range(100):
    a = forward(X, w, b)
    w, b = backward(X, y, a, w, b, lr)

print("Trained weight:", w, "bias:", b)
```

| Concept              | Purpose           | Formula / Logic                             |
|---------------------|-----------------|--------------------------------------------|
| Perceptron           | Binary classifier | $y = f\Big(\sum_i w_i x_i + b\Big)$       |
| Gradient Descent     | Minimize loss     | $w := w - \eta \frac{\partial L}{\partial w}$ |
| Activation Function  | Non-linearity     | Sigmoid, ReLU, Tanh, etc.                  |
| Forward Propagation  | Compute output    | $a^{[l]} = f \Big( W^{[l]} a^{[l-1]} + b^{[l]} \Big)$ |
| Backward Propagation | Update weights    | $W := W - \eta \frac{\partial L}{\partial W}, \quad b := b - \eta \frac{\partial L}{\partial b}$ |


## 🔁 Forward Propagation & Backward Propagation

A neural network learns by repeating these two steps:

- **Forward Propagation** → Make a prediction  
- **Backward Propagation** → Correct mistakes  

This loop continues until the model becomes accurate.

---

### 🟢 1️⃣ Forward Propagation (FP)

Forward propagation means:

**Input → Weights → Activation → Output**

#### 🔹 Mathematical Flow (Single Neuron)

**Step 1: Weighted Sum**


z = w * x + b


**Step 2: Activation**


y = f(z)


Where:  

- \(x\) = input  
- \(w\) = weight  
- \(b\) = bias  
- \(f) = activation function  
- \(y) = predicted output


### 🔹 Example 1: Forward Propagation (Single Neuron)

```python
import numpy as np

# Input
x = np.array([1.0, 2.0])

# Weights & bias
w = np.array([0.5, -1.0])
b = 0.1

# Sigmoid activation
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Forward propagation
z = np.dot(x, w) + b
y_hat = sigmoid(z)

print("Weighted sum (z):", z)
print("Output (ŷ):", y_hat)
```

### 🔴 2️⃣ Loss Function

The network must know **how wrong it is**. This is measured using a **loss function**.

**Example: Mean Squared Error (MSE)**

$$
L = \frac{1}{n} \sum_{i=1}^{n} \left( \hat{y}_i - y_i \right)^2
$$

Where:  

- $L$ = Loss  
- $n$ = Number of samples  
- $y_i$ = Actual output  
- $\hat{y}_i$ = Predicted output



### 🔹 Example 2: Loss Calculation

```python
# True label
y = 1

# Loss
loss = (y - y_hat) ** 2
print("Loss:", loss)
```

### 🔵 3️⃣ Backward Propagation (BP)

Backward propagation means:  

**Adjust weights & bias to reduce loss.**  

It uses:  

- **Chain Rule**  
- **Gradient Descent**

#### 🔹 Why Backpropagation?

- Neural networks have many layers.  
- We must know **how each weight affects the loss**.

#### 🔹 Core Update Rule

For weights:


$$
w := w - \alpha \frac{\partial L}{\partial w}
$$

For bias:

$$
b := b - \alpha \frac{\partial L}{\partial b}
$$

Where:  

- $\alpha$ = learning rate  
- $\frac{\partial L}{\partial w}, \frac{\partial L}{\partial b}$ = gradients of the loss with respect to the weight and bias




### 🧠 Example 3: Forward + Backward (Single Neuron, Manual)

```python
# Learning rate
lr = 0.1

# Derivative of sigmoid
def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)

# Backpropagation
dL_dy = -2 * (y - y_hat)
dy_dz = sigmoid_derivative(z)

dz_dw = x
dz_db = 1

# Gradients
dL_dw = dL_dy * dy_dz * dz_dw
dL_db = dL_dy * dy_dz * dz_db

# Update
w = w - lr * dL_dw
b = b - lr * dL_db

print("Updated weights:", w)
print("Updated bias:", b)
```

✅ This is pure backpropagation math.

### 🟣 Example 4: Training a Neuron (End-to-End)

```python
import numpy as np

# Dataset (OR gate)
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,1])

# Init
w = np.random.randn(2)
b = 0
lr = 0.1
epochs = 5000

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

for _ in range(epochs):
    z = np.dot(X, w) + b
    y_hat = sigmoid(z)

    # Gradients
    dL_dy = y_hat - y
    dy_dz = y_hat * (1 - y_hat)

    dw = np.dot(X.T, dL_dy * dy_dz) / len(X)
    db = np.mean(dL_dy * dy_dz)

    # Update
    w -= lr * dw
    b -= lr * db

print("Predictions:", np.round(sigmoid(np.dot(X, w) + b)))
```

✅ This is a real learning neuron.

## 🧩 Forward vs Backward Propagation (Summary Table)

| Step                  | Purpose                     |
|-----------------------|-----------------------------|
| Forward Propagation    | Predict output              |
| Loss Function          | Measure error               |
| Backward Propagation   | Compute gradients           |
| Gradient Descent       | Update weights              |
| Repeat                 | Learn patterns              |
