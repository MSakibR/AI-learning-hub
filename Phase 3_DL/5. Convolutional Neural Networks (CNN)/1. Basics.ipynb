# 📌 4️⃣ Convolutional Neural Networks (CNN)

---

## 🧠 1️⃣ What is a CNN?

A **CNN** is a type of neural network specially designed for **processing images** (or grid-like data).

- Unlike Dense layers, CNNs take advantage of **spatial structure** (pixels close together are related).

---

### 🔹 Why CNNs?

- Images have **local patterns** (edges, textures, shapes)  
- Dense layers for images → millions of parameters → **slow & overfits easily**  
- CNNs **reduce parameters** with convolutions and pooling

---

## 🧩 2️⃣ CNN Core Components

| Layer                    | Purpose           | Details                          |
|---------------------------|-----------------|---------------------------------|
| Convolutional (Conv2D)   | Extract features | Apply filters/kernels to input  |
| Activation                | Non-linearity    | Usually ReLU                    |
| Pooling (Max/Avg)         | Reduce size      | Downsample feature maps          |
| Flatten                   | Make 1D          | Prepare for Dense layers        |
| Dense / Fully Connected   | Final prediction | Softmax / Sigmoid activation    |

---

### 🔹 Convolution Operation (Mathematics)

For a 2D image:

$$
S(i,j) = (I * K)(i,j) = \sum_m \sum_n I(i+m, j+n) \, K(m,n)
$$

Where:

- $I$ = input image  
- $K$ = kernel / filter  
- $S$ = feature map  

> Each kernel detects patterns like **edges, corners, textures**.

---

## 🟢 3️⃣ Pooling Layer

- Reduces **spatial dimensions**  
- Makes network **translation invariant**  

**Types:**

- Max Pooling → keeps **max value**  
- Average Pooling → keeps **average value**  

**Example:** Max pooling 2x2 → reduces feature map by 4×  

---

## 🔵 4️⃣ Typical CNN Architecture


```python
Input Image → Conv → ReLU → Pool → Conv → ReLU → Pool → Flatten → Dense → Output
```

- Convs + ReLU → learn features
- Pool → reduce dimensions & overfitting  
- Dense → classify images

---

## 🟣 5️⃣ Advantages of CNN

- **Parameter sharing** → fewer weights than Dense layers  
- **Automatic feature extraction** → no need for manual features  
- **Translation invariance** → object recognition even if shifted  
- Works well for **images, video, medical imaging, agriculture**, etc.

---

## 🔹 6️⃣ Example CNN in Keras (Basic MNIST)


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Load MNIST
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train = X_train.reshape(-1,28,28,1)/255.0
X_test = X_test.reshape(-1,28,28,1)/255.0
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)

# CNN Model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D((2,2)),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Train
model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

✅ This is a full CNN for digit recognition.
