# TensorFlow with Keras

**TensorFlow** is the engine, and **Keras** is the easy high-level API built on top of it.  
👉 Think of **Keras = “Deep Learning made simple”**

---

## 🧠 1️⃣ TensorFlow + Keras — Basics

### 🔹 What is TensorFlow?

- A numerical computation and deep learning library developed by **Google**
- Optimized for **GPU / TPU acceleration**
- Widely used in **research and production systems**
- Handles low-level operations like tensors, graphs, and optimization

### 🔹 What is Keras?

- A **high-level API** for building neural networks easily
- Built directly into TensorFlow as `tf.keras`
- Requires **less code**
- Enables **faster development and experimentation**
- Beginner-friendly but powerful enough for production

---

## 🧩 2️⃣ Core Components in Keras

| Component     | Purpose                         |
|--------------|---------------------------------|
| Sequential   | Linear stack of layers          |
| Dense        | Fully connected (FC) layer     |
| Activation   | Adds non-linearity              |
| Loss         | Measures prediction error       |
| Optimizer   | Updates weights & biases        |
| `fit()`      | Trains the model                |
| `predict()`  | Performs inference              |

---



### 🟢 3️⃣ First Keras Model (VERY BASIC)
🔹 Binary Classification (Single Neuron)

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Model
model = Sequential([
    Dense(1, input_shape=(2,), activation='sigmoid')
])

# Compile
model.compile(
    optimizer='sgd',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Data (OR gate)
X = [[0,0],[0,1],[1,0],[1,1]]
y = [0,1,1,1]

# Train
model.fit(X, y, epochs=100, verbose=0)

# Predict
print(model.predict(X).round())
```

✅ This replaces manual forward + backprop automatically.

## 🧠 4️⃣ How Keras Works Internally (Mapping to Theory)

| Theory Concept        | Keras Implementation              |
|----------------------|-----------------------------------|
| Weights              | `Dense` layer                     |
| Activation Function  | `activation='relu'`, `'sigmoid'`  |
| Loss Function        | `loss=`                           |
| Gradient Descent     | `optimizer=`                      |
| Forward Propagation  | Handled automatically             |
| Backpropagation      | Handled automatically             |

---




### 🔵 5️⃣ Multi-Layer Neural Network (MLP)

```python
model = Sequential(
    [
        Dense(16, activation="relu", input_shape=(2,)),
        Dense(8, activation="relu"),
        Dense(1, activation="sigmoid"),
    ]
)

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

model.fit(X, y, epochs=200, verbose=0)
```

✅ Now this is a real deep neural network.

### 🟣 6️⃣ Advanced Keras Example (MNIST)

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load data
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

# Flatten
X_train = X_train.reshape(-1, 28 * 28)
X_test = X_test.reshape(-1, 28 * 28)

# One-hot encode
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Model
model = Sequential(
    [
        Dense(128, activation="relu", input_shape=(784,)),
        Dense(64, activation="relu"),
        Dense(10, activation="softmax"),
    ]
)

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

⭐ This is production-level DL.
