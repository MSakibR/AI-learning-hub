# 🧠 1️⃣ What is an RNN?

RNNs are neural networks designed to handle **sequential data**, where past information affects the future.

**Examples of sequential data:**

- Time series (stock prices, weather)  
- Text / sentences  
- Audio / speech  

Unlike Dense layers, RNNs have **memory**. They pass **hidden states** from previous time steps to the next.

---

### 🔹 Mathematical Formulation

At each time step $t$:

$$
h_t = f(W_{xh} x_t + W_{hh} h_{t-1} + b_h)
$$

$$
y_t = g(W_{hy} h_t + b_y)
$$

Where:

- $x_t$ = input at time $t$  
- $h_t$ = hidden state at time $t$ (memory)  
- $y_t$ = output  
- $W_{xh}, W_{hh}, W_{hy}$ = weight matrices  
- $b_h, b_y$ = biases  
- $f$ = usually **tanh** or **ReLU**  
- $g$ = output activation (**sigmoid / softmax**)  

**Key:** $h_t$ depends on **current input + previous hidden state**

## 🟢 2️⃣ Basic RNN Example in Keras

Predicting a sequence (e.g., sin wave)


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# Generate simple sequential data
t = np.linspace(0, 100, 1000)
data = np.sin(0.1 * t)
X = []
y = []

seq_length = 10
for i in range(len(data) - seq_length):
    X.append(data[i:i+seq_length])
    y.append(data[i+seq_length])

X = np.array(X).reshape(-1, seq_length, 1)
y = np.array(y)

# RNN model
model = Sequential([
    SimpleRNN(50, activation='tanh', input_shape=(seq_length,1)),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.summary()

# Train
model.fit(X, y, epochs=20, batch_size=32)
```

✅ Notes

- `SimpleRNN(50)` → 50 hidden units  
- Input shape = (timesteps, features)  
- Output = next value in sequence

## 🔵 3️⃣ RNN for Text (Sequence Prediction)


```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

sentences = ["I love machine learning", "Deep learning is fun", "RNNs handle sequences"]
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
sequences = tokenizer.texts_to_sequences(sentences)

# pad sequences
X = pad_sequences(sequences, maxlen=4)
y = np.array([1, 0, 1])  # just example labels

# RNN model
model = Sequential([
    SimpleRNN(16, input_shape=(4,1), activation='tanh'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X.reshape(X.shape[0], X.shape[1],1), y, epochs=10)
```

## 🔹 Limitations of RNN

- Vanishing/Exploding Gradients → cannot learn long sequences well

- Cannot remember long-term dependencies

- Training is slower than Dense layers for long sequences

**✅ Solution → LSTM and GRU**
