# GRU (Gated Recurrent Unit)

---

## 🧠 1️⃣ What is GRU?

GRU is a type of RNN that is similar to LSTM but **simpler**:

- Combines forget & input gates into a single **update gate**  
- Has **reset gate** to control past memory contribution  
- Fewer parameters → faster to train  

**✅ Works very well for time-series and sequential data, almost as good as LSTM.**

---

### 🔹 GRU Formulas

At time step $t$:

**Update Gate** – decides how much past info to keep:

$$
z_t = \sigma(W_z \cdot [h_{t-1}, x_t])
$$

**Reset Gate** – decides how much past info to forget:

$$
r_t = \sigma(W_r \cdot [h_{t-1}, x_t])
$$

**Candidate Activation** – new info:

$$
\tilde{h}_t = \tanh(W \cdot [r_t * h_{t-1}, x_t])
$$

**Final Hidden State:**

$$
h_t = (1 - z_t) * h_{t-1} + z_t * \tilde{h}_t
$$

- Simpler than LSTM → no separate cell state, only **hidden state $h_t$**

## 🟢 2️⃣ GRU Example: Time-Series Prediction


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense

# Generate sine wave data
t = np.linspace(0, 100, 1000)
data = np.sin(0.1 * t)

seq_length = 10
X, y = [], []
for i in range(len(data) - seq_length):
    X.append(data[i:i+seq_length])
    y.append(data[i+seq_length])

X = np.array(X).reshape(-1, seq_length, 1)
y = np.array(y)

# GRU Model
model = Sequential([
    GRU(50, activation='tanh', input_shape=(seq_length,1)),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=20, batch_size=32)
```

✅ Notes:

- Works like LSTM for time-series

- Fewer parameters → faster training

## 🔵 3️⃣ GRU Example: Sequential/Text Data

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Embedding

sentences = [
    "I love deep learning",
    "GRU is faster than LSTM",
    "Sequences are fun"
]

# Tokenize
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
sequences = tokenizer.texts_to_sequences(sentences)
X = pad_sequences(sequences, maxlen=5)
y = np.array([1, 0, 1])  # Example labels

# GRU Model
model = Sequential([
    Embedding(input_dim=50, output_dim=8, input_length=5),
    GRU(16),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10)
```

## ✅ Notes

- Input sequences → embedded to vectors  
- GRU handles dependencies in sequences efficiently  
- Output → binary classification  

---

## 🔹 4️⃣ Advantages of GRU over LSTM

| Feature      | LSTM                     | GRU                                |
|-------------|--------------------------|-----------------------------------|
| Gates        | Forget, Input, Output    | Update, Reset                     |
| Cell State   | Yes                      | No (simpler)                      |
| Parameters   | More                     | Fewer → faster training           |
| Performance  | Slightly better on very long sequences | Comparable, faster, less memory |

---

## 🔹 Summary: RNN vs LSTM vs GRU

| Type | Best For         | Memory       | Complexity      | Example Use                        |
|------|-----------------|-------------|----------------|------------------------------------|
| RNN  | Short sequences  | Short-term  | Simple         | Sin wave, short text               |
| LSTM | Long sequences   | Long-term   | More complex   | Stock prediction, NLP              |
| GRU  | Long sequences   | Long-term   | Simpler, faster | Time-series, text classification  |
