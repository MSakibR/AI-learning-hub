# LSTM (Long Short-Term Memory)

---

## 🧠 1️⃣ What is LSTM?

LSTM is a type of RNN designed to **remember information for long sequences**, solving the **vanishing gradient problem** in standard RNNs.

- Introduces **gates** that control information flow: **forget, input, output**

---

### 🔹 LSTM Gates & Formulas

At each time step $t$:

**Forget Gate** – decides what to forget from previous state:

$$
f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f)
$$

**Input Gate** – decides what new information to store:

$$
i_t = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i)
$$

$$
\tilde{C}_t = \tanh(W_C \cdot [h_{t-1}, x_t] + b_C)
$$

**Cell State Update:**

$$
C_t = f_t * C_{t-1} + i_t * \tilde{C}_t
$$

**Output Gate** – decides what to output:

$$
o_t = \sigma(W_o \cdot [h_{t-1}, x_t] + b_o)
$$

$$
h_t = o_t * \tanh(C_t)
$$

Where:

- $C_t$ = cell state (**long-term memory**)  
- $h_t$ = hidden state (**short-term output**)  

**✅ Key:** LSTM can remember **long-term dependencies** better than RNN

## 🟢 2️⃣ LSTM Example: Time-Series Prediction

Predict next value in a sine wave


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Generate sine wave data
t = np.linspace(0, 100, 1000)
data = np.sin(0.1 * t)

seq_length = 10
X, y = [], []
for i in range(len(data) - seq_length):
    X.append(data[i:i+seq_length])
    y.append(data[i+seq_length])

X = np.array(X).reshape(-1, seq_length, 1)
y = np.array(y)

# LSTM Model
model = Sequential([
    LSTM(50, activation='tanh', input_shape=(seq_length, 1)),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=20, batch_size=32)
```

✅ Notes

- LSTM remembers previous 10 time steps  
- Better than RNN for long sequences

## 🔵 3️⃣ LSTM Example: Sequential/Text Data

Predict the next word / sequence classification


```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding

sentences = [
    "I love machine learning",
    "Deep learning is fun",
    "LSTM handles sequences"
]

# Tokenize
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
sequences = tokenizer.texts_to_sequences(sentences)
X = pad_sequences(sequences, maxlen=4)
y = np.array([1, 0, 1])  # Example labels

# LSTM Model
model = Sequential([
    Embedding(input_dim=50, output_dim=8, input_length=4),
    LSTM(16),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10)
```

✅ Notes

- Embedding layer → converts words to vectors  
- LSTM → handles sequential dependencies  
- Output → classification

## 🔹 4️⃣ Advantages of LSTM over RNN

| Feature             | RNN           | LSTM                                  |
|--------------------|---------------|--------------------------------------|
| Memory             | Short-term    | Long-term                             |
| Vanishing Gradient | Yes           | No (solves problem)                   |
| Best for           | Short sequences | Long sequences / Time series / NLP  |
| Gates              | None          | Forget, Input, Output                 |
