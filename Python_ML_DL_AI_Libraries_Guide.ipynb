# 🧠 Python Libraries for ML, DL & AI Projects


## 1️⃣ Core ML & DL Libraries

| Library | What it does | Typical Projects |
|-------|--------------|------------------|
| **NumPy** | Numerical arrays, linear algebra | Almost all ML/DL projects |
| **Pandas** | Dataframes, CSV/Excel handling | Data preprocessing |
| **Scikit-learn** | Traditional ML algorithms | Classification, regression |
| **TensorFlow** | Deep learning framework | CNNs, RNNs, NLP |
| **Keras** | High-level TensorFlow API | Fast DL prototyping |
| **PyTorch** | Flexible DL framework | DL, CV, NLP, RL |



## 2️⃣ Computer Vision Libraries

| Library | Purpose | Projects |

| **OpenCV** | Image/video processing | Real-time CV apps |

## 9️⃣ Supporting Libraries

- **Numba** → Speed up NumPy  
- **tqdm** → Progress bars  
- **joblib / pickle** → Save models  
- **requests / aiohttp** → APIs & datasets  



## 🔹 How to Choose Libraries

- **ML (Tabular)** → NumPy + Pandas + Scikit-learn  
- **DL (CV)** → PyTorch/TensorFlow + OpenCV + YOLO  
- **DL (NLP)** → Transformers + spaCy  
- **Real-Time AI** → OpenCV + MediaPipe + YOLO  
- **Audio AI** → Librosa + SpeechRecognition  



## 💡 Tip
Start with **NumPy, Pandas, OpenCV, PyTorch, Scikit-learn** → then expand.
