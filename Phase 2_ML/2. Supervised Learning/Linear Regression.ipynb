## 🔵 1. What is Linear Regression?

Linear Regression predicts a **continuous numeric value**.

**Example problems:**

- Predict house price  
- Predict student marks  
- Predict salary based on experience  

**The idea:**  
👉 Fit a straight line that best represents the data.

---

## 🔵 2. The Equation of the Line

The model finds:

\[
y = mX + c
\]

Where:

- `m` = slope  
- `c` = intercept  
- `X` = input  
- `y` = predicted output  

In ML terms:

- `m` = weight  
- `c` = bias  

---

## 🔵 3. How the Model Learns?

- The model starts with random weight (`m`) and bias (`c`).  
- Then it tries to **reduce the error** between predicted and real values.

---

## 🔵 4. Cost Function (Most Important)

Linear Regression uses **Mean Squared Error (MSE)**:

\[
\text{Cost} = \frac{1}{n} \sum (y_{\text{pred}} - y_{\text{true}})^2
\]

- Goal → **Minimize this cost**.

---

## 🔵 5. Gradient Descent (Concept)

To reduce error, the model adjusts `m` and `c` slightly:

1. Compute error  
2. Adjust `m` a bit  
3. Adjust `c` a bit  
4. Repeat thousands of times  

Eventually, **cost becomes minimum**.  
This is how the model “learns”.

---

## 🔵 6. Implement Linear Regression (sklearn)

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Dataset: Study hours vs Marks
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([40, 50, 60, 70, 80])

model = LinearRegression()
model.fit(X, y)

# Predict value
print(model.predict([[6]]))  # ~90 marks

# Plot
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.show()


### 7. Check Model Parameters

````python
print("Slope (m):", model.coef_)
print("Intercept (c):", model.intercept_)
````

### 8. Evaluate the Model

````python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X)

print("MSE:", mean_squared_error(y, y_pred))
print("R2 Score:", r2_score(y, y_pred))
````

#### Final

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression  # sklearn - Scikit-learn
from sklearn.metrics import mean_squared_error, r2_score

# Dataset: Study hours vs Marks
X = np.array([[1], [2], [3], [4], [5]])  # 2D (feature)
y = np.array([40, 50, 60, 70, 80])  # 1D (target)

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict marks for 6 hours study
print("Prediction for 6 hours:", model.predict([[6]])[0])

# Plot data points
plt.scatter(X, y, label="Actual Data")

# Plot regression line
y_line = model.predict(X)
plt.plot(X, y_line, color="red", label="Best Fit Line")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression Example")
plt.legend()
plt.show()

# Print slope & intercept
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)

# Evaluate model
y_pred = model.predict(X)
print("MSE:", mean_squared_error(y, y_pred))
print("R2 Score:", r2_score(y, y_pred))
````
