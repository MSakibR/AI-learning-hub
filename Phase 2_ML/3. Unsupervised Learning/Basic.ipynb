# 🌟 Unsupervised Learning — Basic Idea

Unlike Supervised Learning, where we have features (`X`) and labels (`y`), in **Unsupervised Learning** we only have features.  

- There are **no labels / no target variable**  
- The goal is to **find patterns, structures, or relationships** in the data  

---

## 🔹 Key Tasks in Unsupervised Learning

- **Clustering →** Group similar data points together  
  - *Example:* Segmenting customers into groups  

- **Dimensionality Reduction →** Reduce features while keeping important info  
  - *Example:* Visualizing high-dimensional data, speeding up ML models  

---

## 🔹 Key Differences Between Supervised & Unsupervised Learning

| Feature         | Supervised Learning          | Unsupervised Learning              |
|-----------------|----------------------------|----------------------------------|
| Labels          | Required                   | Not required                     |
| Goal            | Predict target variable     | Find hidden patterns             |
| Examples        | Regression, Classification | Clustering, PCA                  |
| Evaluation      | Accuracy, MSE              | Silhouette score, explained variance |

---

## 🔹 Why Use Unsupervised Learning?

- **Explore the data →** find hidden patterns before building models  
- **Reduce dimensions →** easier visualization, faster ML  
- **Segment or group data →** marketing, recommendation systems, anomaly detection  

---

## 💡 Real-Life Examples

- **Customer Segmentation:** Grouping users based on buying behavior  
- **Anomaly Detection:** Detecting fraudulent transactions  
- **Dimensionality Reduction:** PCA for facial recognition or image compression  
- **Market Basket Analysis:** Find product bundles
