# 📊 PCA — Principal Component Analysis (Dimensionality Reduction)

---

## 🔵 1️⃣ What is PCA?

- PCA is a technique to **reduce the number of features (dimensions)** while keeping most of the important information.  
- It transforms original features into new **principal components** that are **uncorrelated**.  
- **Analogy:** Think of it as compressing data without losing much information.  

---

## 🔵 2️⃣ Why Use PCA?

- High-dimensional data → hard to visualize or train models  
- Reduces computational cost  
- Removes redundancy (correlated features)  
- Improves performance of ML models  

**Example:** 50 features → PCA → 2–3 components for visualization  

---

## 🔵 3️⃣ How PCA Works (Intuition)

1. Compute **covariance matrix** of the data  
2. Find **eigenvectors and eigenvalues** → directions of maximum variance  
3. Sort eigenvectors by eigenvalues → top eigenvectors = principal components  
4. Project original data onto **principal components**  

- The **first component** captures the most variance  
- The **second component** captures the next most, and so on  

---

## 🔵 4️⃣ Python Example — PCA with Iris Dataset



```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load data
data = load_iris()
X = data.data
y = data.target

# Apply PCA (2 components for visualization)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.show()

# Explained variance
print("Explained variance ratio:", pca.explained_variance_ratio_)

#✅ You’ll see a 2D plot separating Iris species based on PCA components.
```

## 🔵 5️⃣ Key Points

- `n_components` → number of principal components to keep  
- `explained_variance_ratio_` → how much variance each component captures  
- PCA does **not consider labels** → purely unsupervised  
- Can be used **before ML algorithms** to reduce dimensionality  

---

## 🔵 6️⃣ Advantages

- ✔ Reduces dimensionality → faster computation  
- ✔ Removes correlated/redundant features  
- ✔ Good for visualization  
- ✔ Improves model performance in some cases  

---

## 🔵 7️⃣ Disadvantages

- ❌ Can lose some information (if too few components)  
- ❌ Components are linear combinations → hard to interpret  
- ❌ Sensitive to feature scaling → always scale before PCA

```python
# ==========================
# PCA Mini Practice: Wine Dataset
# ==========================
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# 1️⃣ Load Wine dataset
data = load_wine()
X = data.data
y = data.target

print(f"Original data shape: {X.shape}")  # 13 features

# 2️⃣ Reduce to 2 components for visualization
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X)

# Plot clusters
plt.figure(figsize=(6, 5))
plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=y, cmap="viridis", s=50)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("2D PCA of Wine Dataset")
plt.show()

# 3️⃣ Explained variance
print("Explained variance ratio (2D):", pca_2d.explained_variance_ratio_)
print("Total variance preserved (2D):", sum(pca_2d.explained_variance_ratio_))

# 4️⃣ Optional: Try 3 components → train ML model
pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X)
print("Explained variance ratio (3D):", pca_3d.explained_variance_ratio_)
print("Total variance preserved (3D):", sum(pca_3d.explained_variance_ratio_))

# Split data and train model on 3D PCA data
X_train, X_test, y_train, y_test = train_test_split(
    X_pca_3d, y, test_size=0.2, random_state=42
)
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("RandomForest Accuracy on 3D PCA data:", acc)
```

✅ Explanation

- Load Wine Dataset → 13 chemical features, 3 wine classes.

- PCA 2D → reduce 13 features to 2 for visualization.

- Scatter plot → shows how well PCA separates the wine classes.

- Explained variance ratio → tells how much information is preserved:

    2D may preserve ~60–70% of variance

    3D preserves more (~80–90%)

- Train ML model (RandomForest) on PCA-reduced data → compare accuracy:

  Fewer components → faster training

  More components → potentially higher accuracy

```python
# ==========================
# 3D PCA on Iris
# ==========================
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Load data
data = load_iris()
X = data.data
y = data.target
target_names = data.target_names

# Apply PCA (3 components)
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

# Explained variance
print("Explained variance ratio (3D):", pca.explained_variance_ratio_)
print("Total variance preserved:", sum(pca.explained_variance_ratio_))

# 3D Scatter plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")

colors = ["r", "g", "b"]
for target, color, name in zip([0, 1, 2], colors, target_names):
    ax.scatter(
        X_pca[y == target, 0],
        X_pca[y == target, 1],
        X_pca[y == target, 2],
        c=color,
        label=name,
        s=50,
    )

ax.set_xlabel("PC 1")
ax.set_ylabel("PC 2")
ax.set_zlabel("PC 3")
ax.set_title("3D PCA of Iris Dataset")
ax.legend()
plt.show()
```

✅ What this code does

1. PCA with 3 components → reduces 4D Iris features to 3D.

2. Explained variance ratio → shows how much info is preserved:

  - Typically, PC1 + PC2 + PC3 ≈ 95–98% of the variance.

3. 3D scatter plot → each species is colored differently:

  - setosa → red

  - versicolor → green

  - virginica → blue

4. Axes: PC1, PC2, PC3 → new features capturing maximum variance.
