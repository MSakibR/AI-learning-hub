# 🌳 Hierarchical Clustering

Unlike K-Means, **Hierarchical Clustering** doesn’t require you to specify the number of clusters upfront.  

---

## 🔵 1️⃣ What is Hierarchical Clustering?

- Builds a **tree-like structure (dendrogram)** to show data hierarchy  
- Two types:  
  - **Agglomerative (Bottom-Up):** Start with each point as its own cluster, merge closest clusters step by step  
  - **Divisive (Top-Down):** Start with all points in one cluster, split recursively  
- Most commonly used: **Agglomerative**

---

## 🔵 2️⃣ How Agglomerative Clustering Works

1. Treat each data point as a cluster  
2. Compute distance between clusters (Euclidean / Manhattan)  
3. Merge the closest two clusters  
4. Repeat until all points belong to one cluster  
5. Visualize with **dendrogram** → cut at desired level to get clusters  

---

## 🔵 3️⃣ Linkage Methods (Distance Between Clusters)

| Method   | Description                                   |
|----------|-----------------------------------------------|
| Single   | Distance between closest points of clusters  |
| Complete | Distance between farthest points of clusters |
| Average  | Average distance between all points          |
| Ward     | Minimizes variance within clusters (most popular) |

---

## 🔵 4️⃣ Python Example (Agglomerative Clustering)




```python
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Generate sample data
X, _ = make_blobs(n_samples=50, centers=3, random_state=42)

# Compute linkage matrix for dendrogram
linked = linkage(X, method='ward')

# Plot dendrogram
plt.figure(figsize=(8,4))
dendrogram(linked, orientation='top', distance_sort='ascending', show_leaf_counts=True)
plt.show()

# Create clusters
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
labels = cluster.fit_predict(X)

# Plot clusters
plt.scatter(X[:,0], X[:,1], c=labels, cmap='rainbow')
plt.show()
```

## 🔵 5️⃣ Advantages

- ✔ No need to specify number of clusters upfront  
- ✔ Produces a **dendrogram** → visualize hierarchy  
- ✔ Works well with small datasets  

---

## 🔵 6️⃣ Disadvantages

- ❌ Computationally expensive for large datasets  
- ❌ Sensitive to noise and outliers  
- ❌ Hard to scale

```python
# ==========================
# Hierarchical Clustering Mini Task
# ==========================
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# 1️⃣ Generate sample data
X, _ = make_blobs(n_samples=60, centers=3, cluster_std=0.6, random_state=42)

# 2️⃣ Plot dendrogram to decide number of clusters
plt.figure(figsize=(8,4))
linked = linkage(X, method='ward')  # You can try 'single', 'complete', 'average'
dendrogram(linked, orientation='top', distance_sort='ascending', show_leaf_counts=True)
plt.title("Dendrogram (ward linkage)")
plt.show()

# 3️⃣ Fit AgglomerativeClustering and visualize clusters
n_clusters = 3  # Decide based on dendrogram
cluster = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage='ward')
labels = cluster.fit_predict(X)

plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='rainbow', s=50)
plt.title(f"Agglomerative Clustering with {n_clusters} clusters (ward linkage)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# 4️⃣ Experiment with different linkage methods
linkage_methods = ['single', 'complete', 'average']

for method in linkage_methods:
    cluster = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage=method)
    labels = cluster.fit_predict(X)
    
    plt.figure(figsize=(6,5))
    plt.scatter(X[:,0], X[:,1], c=labels, cmap='rainbow', s=50)
    plt.title(f"Agglomerative Clustering ({method} linkage)")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```

✅ What this code does

- Generate 2D blobs with make_blobs(n_samples=60, centers=3) → synthetic clusters.\
- Plot dendrogram → helps you visually decide how many clusters to choose.\
- Fit AgglomerativeClustering with n_clusters=3 and plot the clusters.\
- Experiment with different linkage methods (single, complete, average) → see how cluster shapes change.
